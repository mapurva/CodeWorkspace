# Intuition Behind Hypothesis Testing

Before learning formulas, let's understand **why hypothesis testing is needed**.

---

## Test Case 1: Is the Professor Late Today?

Imagine this situation.

Your class normally starts at **10:00 AM**.

But students have observed something over the semester:
- The professor usually arrives around **10:00 AM**
- Sometimes **1–2 minutes late**
- Rarely **more than 5 minutes late**

Today the professor arrives at **10:07 AM**.

Now students start discussing:

- Student A: "Maybe this is normal variation."
- Student B: "No, this is unusual. Something must have happened."

So the real question becomes:

> Is **10:07 AM** still part of normal variation, or is it **significantly different** from usual behavior?

---

## What Hypothesis Testing Does

Hypothesis testing gives a **systematic method** to answer this question.

We start with an assumption:

**Null Hypothesis (H₀):**
> Nothing unusual happened. The professor is arriving at the usual time.

Then we ask:

> If the professor normally arrives around 10:00 AM,  
> how likely is it to observe **10:07 AM**?

If the probability is **very small**, we say:

➡ Something unusual happened  
➡ Reject the null hypothesis

---

### Key Intuition

Hypothesis testing answers a simple question:

> **Is the observed result just random variation, or is it evidence of something new?**

# What is Hypothesis Testing?

Hypothesis testing is a **statistical method used to make decisions using data**.

In many real-world situations, we observe some data and want to answer a question:

> Is this result due to **random variation**, or does it indicate **something significant**?

Hypothesis testing provides a **structured framework** to answer this question.

---

# Basic Idea

The process begins with an assumption about the population.

This assumption is called the **Null Hypothesis**.

### Null Hypothesis (H₀)

The null hypothesis assumes that:

> **Nothing unusual is happening.**

It represents the **status quo** or the **default belief**.

Example:
- A training program does not improve productivity
- A new teaching method does not improve scores
- The average packet weight is exactly 50g

---

### Alternative Hypothesis (H₁ or Hₐ)

The alternative hypothesis represents the **claim we want to test**.

It suggests that:

> **Something has changed or there is a real effect.**

Examples:
- The training program increases productivity
- The new teaching method improves scores
- The packet weight is different from 50g

---

# What Hypothesis Testing Actually Does

Hypothesis testing evaluates the following question:

> If the null hypothesis is true, how likely is the observed data?

Two possibilities arise:

### Case 1: Data is Likely Under H₀
If the observed data is **reasonably likely**, we say:

➡ **Fail to reject the null hypothesis**

This means:
- There is **not enough evidence** to support the alternative hypothesis.

---

### Case 2: Data is Very Unlikely Under H₀
If the observed data is **very unlikely**, we say:

➡ **Reject the null hypothesis**

This means:
- There is **strong evidence** supporting the alternative hypothesis.

---

If evidence is strong enough:

If evidence is not strong enough:

# Important Points About Hypothesis Testing

Before performing any hypothesis test, it is important to understand some key principles.

---

## 1. Null Hypothesis Represents the Status Quo

The **Null Hypothesis (H₀)** usually represents the assumption that:

> Nothing new is happening.

It assumes that there is **no significant effect or relationship** between the variables being studied.

Examples:
- A new teaching method does **not** improve scores.
- A training program does **not** increase productivity.
- The average packet weight is **50 grams**.

---

## 2. Alternative Hypothesis Represents the Research Claim

The **Alternative Hypothesis (H₁ or Hₐ)** contradicts the null hypothesis.

It represents the claim that:

> There **is a significant effect or difference**.

Examples:
- The new teaching method **improves scores**.
- The training program **increases productivity**.
- The average packet weight is **different from 50 grams**.

---

## 3. We Try to Find Evidence Against the Null Hypothesis

Hypothesis testing works by using **sample data** to evaluate whether the null hypothesis is reasonable.

The goal is to determine:

> Is the observed result likely to occur if the null hypothesis were true?

If the observed data is **very unlikely under H₀**, we reject the null hypothesis.

---

## 4. Two Possible Decisions

After performing a hypothesis test, there are only two possible outcomes:

1️⃣ **Reject the Null Hypothesis (H₀)**  
This means the sample provides **sufficient evidence** supporting the alternative hypothesis.

2️⃣ **Fail to Reject the Null Hypothesis (H₀)**  
This means the sample does **not provide enough evidence** to support the alternative hypothesis.

---

## 5. Failing to Reject H₀ Does NOT Mean H₀ is True

A very important point:

> **Failing to reject the null hypothesis does not mean the null hypothesis is true.**

It only means that **the available evidence is insufficient to reject it**.

This happens because conclusions are based on **sample data**, which always contains some uncertainty.

---

## 6. Hypothesis Testing is Evidence-Based Decision Making

Hypothesis testing helps answer the question:

> Is the observed result just **random variation**, or is it **statistically significant evidence** of a real effect?

By using statistical tests, we make **data-driven decisions instead of relying on intuition alone**.

# Steps Involved in Hypothesis Testing (Rejection Region Approach)

Hypothesis testing follows a systematic procedure to decide whether the **sample data provides enough evidence to reject the null hypothesis**.

Below are the steps involved in the **Rejection Region Approach**.

---

## Step 1: Formulate the Hypotheses

Define two hypotheses:

**Null Hypothesis (H₀):**  
Represents the default assumption or status quo.  
It usually states that **there is no effect or no difference**.

**Alternative Hypothesis (H₁ or Hₐ):**  
Represents the research claim.  
It states that **there is an effect or a difference**.

Example:  
H₀: μ = 50  
H₁: μ ≠ 50

---

## Step 2: Select the Significance Level (α)

Choose a significance level **α (alpha)**, which represents the probability of committing a **Type I error**.

A **Type I error** occurs when we **reject the null hypothesis even though it is actually true**.

Common values for α:
- 0.05 (5%)
- 0.01 (1%)

---

## Step 3: Check Assumptions

Before performing the test, verify the assumptions required for the chosen statistical test.

Examples:
- Data distribution (often normal distribution)
- Independence of observations
- Sample size requirements

---

## Step 4: Choose the Appropriate Statistical Test

Depending on the problem and data characteristics, choose the appropriate test.

Examples:
- **Z-test** → when population standard deviation is known
- **t-test** → when population standard deviation is unknown
- **Chi-square test** → for categorical data
- **ANOVA** → for comparing means of multiple groups

---

## Step 5: Determine the Test Statistic and Critical Region

Calculate the **test statistic** based on the sample data.

Example (Z-test statistic):

Z = (X̄ − μ) / (σ / √n)

Using the chosen significance level (α), determine the **critical value(s)** from statistical tables.

These critical values define the **rejection region**.

If the test statistic falls in this region, the null hypothesis will be rejected.

---

## Step 6: Conduct the Test

Compute the value of the **test statistic** using the sample data.

Then compare this value with the **critical value(s)** determined earlier.

---

## Step 7: Make the Decision

Two outcomes are possible:

If the test statistic lies in the **rejection region**

➡ Reject the null hypothesis (H₀)

If the test statistic lies **outside the rejection region**

➡ Fail to reject the null hypothesis (H₀)

---

## Step 8: Interpret the Result

Finally, interpret the statistical conclusion in the context of the problem.

Example conclusion:

"The sample provides sufficient evidence to conclude that the training program significantly increases productivity."

or

"There is insufficient evidence to conclude that the average weight differs from 50 grams."

# Performing a Z-Test: Example 1

A company wants to evaluate whether a **new training program improves employee productivity**.

---

## Step 1: Problem Description

Before the training program:

- Average productivity = **50 units per day**

The population standard deviation is known:

- **σ = 5 units**

After implementing the training program, the company measures productivity for a **random sample of 30 employees**.

The sample statistics are:

- Sample size (n) = **30**
- Sample mean (x̄) = **53 units**

The question is:

> Has the training program significantly increased productivity?

---

## Step 2: State the Hypotheses

Since the company wants to check whether productivity **increased**, we use a **one-tailed test**.

**Null Hypothesis (H₀):**  
μ = 50  
(The training program does not increase productivity.)

**Alternative Hypothesis (H₁):**  
μ > 50  
(The training program increases productivity.)

---

## Step 3: Select the Significance Level

Assume the significance level is:

α = **0.05**

This means we allow a **5% chance of rejecting the null hypothesis when it is actually true**.

---

## Step 4: Choose the Test Statistic

Since the **population standard deviation (σ) is known**, we use a **Z-test**.

The Z-test statistic is: Z = (x̄ − μ) / (σ / √n)


where:

- x̄ = sample mean  
- μ = population mean  
- σ = population standard deviation  
- n = sample size  

---

## Step 5: Compute the Test Statistic

Substitute the values: Z = (53 − 50) / (5 / √30)

First calculate the denominator: 5 / √30 ≈ 5 / 5.477 ≈ 0.912

Now compute Z: Z = 3 / 0.912 ≈ 3.29


So the **test statistic is**

Z ≈ **3.29**

---

## Step 6: Determine the Critical Value

For a **right-tailed test** with α = 0.05,
https://www.ztable.net/
Critical Z value: Z₍critical₎ = 1.645


Decision rule:

- If **Z > 1.645 → Reject H₀**
- Otherwise → Fail to reject H₀

---

## Step 7: Make the Decision

Calculated Z value: Z = 3.29


Since: 3.29 > 1.645


The test statistic falls in the **rejection region**.

Therefore:

➡ **Reject the null hypothesis (H₀)**

---

## Step 8: Interpretation

There is **sufficient statistical evidence** at the 5% significance level to conclude that:

> The new training program significantly **increases employee productivity**.

# Visualizing the Rejection Region for the Z-Test

From the previous example we computed the test statistic:

Z = **3.29**

We also determined the **critical value** for a right-tailed test with α = 0.05:

Z₍critical₎ = **1.645**

Decision rule:

- If **Z > 1.645 → Reject H₀**
- If **Z ≤ 1.645 → Fail to reject H₀**

---

## Interpretation

The rejection region is located in the **right tail of the normal distribution**.

This region represents values that are **very unlikely if the null hypothesis is true**.

Since our calculated value:

Z = **3.29**

lies **far into the rejection region**, we reject the null hypothesis.

---

## What the Plot Will Show

In the visualization:

- The **blue curve** represents the normal distribution under H₀
- The **red shaded region** shows the rejection region
- The **vertical line** represents the calculated Z statistic (3.29)

This helps visually understand why we reject the null hypothesis.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Z values
z = np.linspace(-4, 4, 500)

# normal distribution
y = norm.pdf(z)

# critical value and test statistic
z_critical = 1.645
z_stat = 3.29

plt.figure(figsize=(8,5))
plt.plot(z, y, label="Normal Distribution")

# rejection region
z_reject = np.linspace(z_critical, 4, 200)
plt.fill_between(z_reject, norm.pdf(z_reject), alpha=0.3, label="Rejection Region (α = 0.05)")

# vertical lines
plt.axvline(z_critical, linestyle="--", label="Critical Value (1.645)")
plt.axvline(z_stat, linestyle=":", label="Test Statistic (3.29)")

plt.xlabel("Z value")
plt.ylabel("Probability Density")
plt.title("Rejection Region for Z-Test")
plt.legend()
plt.show()
````

# Performing a Z-Test: Example 2

A snack food company claims that their **Lays wafer packets contain an average weight of 50 grams per packet**.

A consumer watchdog organization wants to verify this claim.

---

## Step 1: Problem Description

The organization collects a **random sample of 40 packets** and measures their weights.

The sample statistics are:

- Claimed population mean (μ) = **50 grams**
- Sample mean (x̄) = **49 grams**
- Population standard deviation (σ) = **4 grams**
- Sample size (n) = **40**

The question is:

> Does the actual average packet weight differ from the claimed 50 grams?

---

## Step 2: State the Hypotheses

Since we want to check whether the average weight is **different** from 50 grams, we use a **two-tailed test**.

**Null Hypothesis (H₀):**  
μ = 50  
(The average packet weight is 50 grams.)

**Alternative Hypothesis (H₁):**  
μ ≠ 50  
(The average packet weight is different from 50 grams.)

---

## Step 3: Select the Significance Level

Assume the significance level:

α = **0.05**

Since this is a **two-tailed test**, the significance level is divided between both tails:

α/2 = **0.025** in each tail.

---

## Step 4: Choose the Test Statistic

Since the **population standard deviation (σ) is known**, we use a **Z-test**.

The test statistic formula is:

Z = (x̄ − μ) / (σ / √n)

where:

- x̄ = sample mean  
- μ = population mean  
- σ = population standard deviation  
- n = sample size  

---

## Step 5: Compute the Test Statistic

Substitute the values:

Z = (49 − 50) / (4 / √40)

First compute the denominator:

√40 ≈ 6.324  

4 / 6.324 ≈ **0.632**

Now compute Z:

Z = −1 / 0.632 ≈ **−1.58**

So the **test statistic is**

Z ≈ **−1.58**

---

## Step 6: Determine the Critical Values

For a **two-tailed test** with α = 0.05:
https://www.ztable.net/
Critical Z values are:

Z₍critical₎ = **±1.96**

Decision rule:

- If **Z < −1.96** or **Z > 1.96 → Reject H₀**
- Otherwise → **Fail to reject H₀**

---

## Step 7: Make the Decision

Calculated Z value:

Z = **−1.58**

Since:

−1.96 < −1.58 < 1.96

The test statistic lies **within the acceptance region**.

Therefore:

➡ **Fail to reject the null hypothesis (H₀)**

---

## Step 8: Interpretation

At the **5% significance level**, there is **not enough statistical evidence** to conclude that the average packet weight differs from **50 grams**.

This means the observed difference in the sample could be due to **random variation**.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate Z values
z = np.linspace(-4, 4, 500)
y = norm.pdf(z)

# Critical values and test statistic
z_critical_left = -1.96
z_critical_right = 1.96
z_stat = -1.58

plt.figure(figsize=(8,5))

# Plot normal distribution
plt.plot(z, y, label="Normal Distribution")

# Left rejection region
z_left = np.linspace(-4, z_critical_left, 200)
plt.fill_between(z_left, norm.pdf(z_left), alpha=0.3, label="Left Rejection Region (α/2 = 0.025)")

# Right rejection region
z_right = np.linspace(z_critical_right, 4, 200)
plt.fill_between(z_right, norm.pdf(z_right), alpha=0.3, label="Right Rejection Region (α/2 = 0.025)")

# Critical value lines
plt.axvline(z_critical_left, linestyle="--", label="Critical Value -1.96")
plt.axvline(z_critical_right, linestyle="--", label="Critical Value 1.96")

# Test statistic line
plt.axvline(z_stat, linestyle=":", label="Test Statistic Z = -1.58")

plt.xlabel("Z value")
plt.ylabel("Probability Density")
plt.title("Two-Tailed Z-Test Rejection Regions (α = 0.05)")
plt.legend()
plt.show()
````

# Rejection Region Approach in Hypothesis Testing

One way to perform hypothesis testing is the **Rejection Region Approach** (also called the **Critical Value Approach**).

This method decides whether to reject the null hypothesis by comparing the **test statistic** with a **critical value** determined from the probability distribution.

---

# Significance Level (α)

The **significance level**, denoted by **α (alpha)**, is a threshold used to decide whether the null hypothesis should be rejected.

It represents the probability of making a **Type I error**.

Type I Error:
> Rejecting the null hypothesis when it is actually true.

Typical values of α used in practice:

- α = 0.05 (5%)
- α = 0.01 (1%)

For example:

If α = 0.05, we accept a **5% chance of incorrectly rejecting the null hypothesis**.

---

# What is the Rejection Region?

The **Rejection Region (Critical Region)** is the set of values of the test statistic that lead to the rejection of the null hypothesis.

These values correspond to outcomes that are **very unlikely if the null hypothesis is true**.

Decision rule:

- If the test statistic falls in the **rejection region → Reject H₀**
- Otherwise → **Fail to reject H₀**

---

# Visual Intuition

In hypothesis testing, the rejection region lies in the **extreme tails of the probability distribution**.

For example, in a **two-tailed test with α = 0.05**:

- 2.5% probability lies in the **left tail**
- 2.5% probability lies in the **right tail**

These tails represent **rare events under the null hypothesis**.

If the test statistic falls in these extreme regions, we conclude that the observed result is **too unlikely to occur if H₀ were true**.

---

# Steps in the Rejection Region Approach

1. State the null and alternative hypotheses.
2. Choose the significance level (α).
3. Select the appropriate statistical test.
4. Determine the **critical value(s)** from statistical tables.
5. Compute the **test statistic** using sample data.
6. Compare the test statistic with the **critical value(s)**.
7. Decide whether to **reject or fail to reject H₀**.

---

# Limitation of the Rejection Region Approach

Although the rejection region method is useful for understanding hypothesis testing, it has some limitations.

### 1. Only Gives a Yes/No Decision

It only tells us whether to:

- Reject H₀
- Fail to reject H₀

It does **not indicate how strong the evidence is** against the null hypothesis.

---

### 2. Requires Looking Up Critical Values

We must use **statistical tables** to find critical values, which can be inconvenient.

---

### 3. Less Informative for Interpretation

The rejection region approach does not tell us:

> *How unlikely the observed result actually is.*

---

# Why We Use the p-Value Approach

To overcome these limitations, statisticians often use the **p-value approach**.

The **p-value** is defined as:

> The probability of obtaining a result **at least as extreme as the observed result**, assuming the null hypothesis is true.

Instead of comparing with a critical value, we compare the **p-value with α**.

Decision rule:

- If **p-value ≤ α → Reject H₀**
- If **p-value > α → Fail to reject H₀**

---

# Advantage of the p-Value Approach

The p-value provides **more information** than the rejection region approach.

For example:

- p-value = 0.04 → weak evidence against H₀
- p-value = 0.001 → very strong evidence against H₀

Thus, the p-value tells us **how strong the evidence is** against the null hypothesis.

---

# Key Takeaway

The **Rejection Region Approach** helps us understand the logic of hypothesis testing, while the **p-value approach** provides a more flexible and informative way to make statistical decisions.

# Visual Insight: Rejection Region vs p-value

To better understand hypothesis testing, we can visualize the **normal distribution under the null hypothesis (H₀)**.

Key ideas:

- The **center of the distribution** represents expected outcomes if H₀ is true.
- The **extreme tails** represent rare outcomes.

---

## Significance Level (α)

The **significance level α** defines the **rejection region**.

Example:

If α = 0.05 (two-tailed test)

- 2.5% probability lies in the **left tail**
- 2.5% probability lies in the **right tail**

These tail regions are the **rejection regions**.

If the test statistic falls in these regions:

➡ Reject H₀

---

## What the Visualization Shows

In the plot:

- Blue curve → sampling distribution under H₀
- Red shaded areas → **rejection regions (α)**
- Vertical dashed lines → **critical values**
- Black vertical line → **observed test statistic**

---

## Connecting to the p-value

The **p-value** measures how extreme the observed result is.

It represents:

> The probability of observing a result **as extreme or more extreme** than the test statistic assuming H₀ is true.

Unlike the rejection region approach, the p-value tells us **how strong the evidence is**.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate values
x = np.linspace(-4, 4, 500)
y = norm.pdf(x)

# Parameters
alpha = 0.05
z_critical = 1.96   # two-tailed critical value
z_stat = 2.5        # example observed statistic

plt.figure(figsize=(8,5))

# Plot distribution
plt.plot(x, y, label="Normal Distribution (H₀ true)")

# Rejection regions
left_tail = np.linspace(-4, -z_critical, 200)
right_tail = np.linspace(z_critical, 4, 200)

plt.fill_between(left_tail, norm.pdf(left_tail), alpha=0.3,
                 label="Rejection Region (α/2)")
plt.fill_between(right_tail, norm.pdf(right_tail), alpha=0.3)

# Critical values
plt.axvline(-z_critical, linestyle="--", label="Critical Values ±1.96")
plt.axvline(z_critical, linestyle="--")

# Test statistic
plt.axvline(z_stat, linestyle=":", label=f"Test Statistic Z = {z_stat}")

plt.xlabel("Z value")
plt.ylabel("Probability Density")
plt.title("Rejection Region in Hypothesis Testing")
plt.legend()
plt.show()

# Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print("p-value:", p_value)
````

# Type I Error vs Type II Error

In hypothesis testing, decisions are made based on **sample data**.  
Because sample data contains uncertainty, two types of errors can occur.

---

# Type I Error (False Positive)

A **Type I error** occurs when we:

> Reject the null hypothesis even though it is actually true.

In other words, we detect an effect **when no real effect exists**.

The probability of committing a Type I error is denoted by:

α (alpha)

This is also called the **significance level**.

Example:

- H₀: A new drug has **no effect**
- H₁: The drug **improves recovery**

If we reject H₀ when the drug actually has **no effect**,  
we have committed a **Type I error**.

---

# Type II Error (False Negative)

A **Type II error** occurs when we:

> Fail to reject the null hypothesis even though it is false.

This means we **miss detecting a real effect**.

The probability of committing a Type II error is denoted by:

β (beta)

Example:

- H₀: A new drug has **no effect**
- H₁: The drug **improves recovery**

If the drug actually works but our test **fails to detect it**,  
we have committed a **Type II error**.

---

# Summary of Possible Outcomes

| Reality | Decision | Result |
|------|------|------|
| H₀ true | Fail to reject H₀ | Correct decision |
| H₀ true | Reject H₀ | Type I Error |
| H₀ false | Reject H₀ | Correct decision |
| H₀ false | Fail to reject H₀ | Type II Error |

---

# Trade-off Between Type I and Type II Errors

Reducing one error often increases the other.

Example:

If we choose a **very small significance level (α = 0.01)**

- Type I error becomes **less likely**
- But Type II error becomes **more likely**

Therefore, hypothesis testing involves **balancing these two risks**.

---

# Key Intuition

Type I error → **False alarm**

Type II error → **Missed detection**

Understanding these errors is important when designing statistical tests and interpreting results.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define distributions
x = np.linspace(-4, 6, 500)

# Distribution when H0 is true
y_null = norm.pdf(x, 0, 1)

# Distribution when H1 is true
y_alt = norm.pdf(x, 2, 1)

# Critical value
z_critical = 1.645

plt.figure(figsize=(9,5))

# Plot distributions
plt.plot(x, y_null, label="Distribution when H₀ is true")
plt.plot(x, y_alt, label="Distribution when H₁ is true")

# Type I error region
x_type1 = np.linspace(z_critical, 6, 200)
plt.fill_between(x_type1, norm.pdf(x_type1,0,1), alpha=0.3,
                 label="Type I Error (α)")

# Type II error region
x_type2 = np.linspace(-4, z_critical, 200)
plt.fill_between(x_type2, norm.pdf(x_type2,2,1), alpha=0.3,
                 label="Type II Error (β)")

# Critical value
plt.axvline(z_critical, linestyle="--", label="Critical Value")

plt.xlabel("Test Statistic")
plt.ylabel("Probability Density")
plt.title("Type I vs Type II Errors in Hypothesis Testing")
plt.legend()
plt.show()
````

# One-Tailed vs Two-Tailed Tests

In hypothesis testing, the type of test depends on the **form of the alternative hypothesis (H₁)**.

The alternative hypothesis determines whether the test is:

- **One-tailed (one-sided) test**
- **Two-tailed (two-sided) test**

---

# One-Tailed Test

A **one-tailed test** is used when we are interested in detecting a change in **only one direction**.

The alternative hypothesis contains either:

- **greater than (>)**, or
- **less than (<)**

Examples:

Right-tailed test  
H₀: μ = 50  
H₁: μ > 50  

Example interpretation:
> Testing whether a new training program **increases productivity**.

Left-tailed test  
H₀: μ = 50  
H₁: μ < 50  

Example interpretation:
> Testing whether a manufacturing process **reduces product weight**.

---

# Two-Tailed Test

A **two-tailed test** is used when we want to detect a change in **either direction**.

The alternative hypothesis contains:

≠ (not equal to)

Example:

H₀: μ = 50  
H₁: μ ≠ 50  

Example interpretation:
> Testing whether the average packet weight **differs from 50 grams**.

---

# Distribution of Significance Level

The main difference between these tests lies in **how the significance level (α) is allocated**.

### One-Tailed Test

All of α is placed in **one tail** of the distribution.

Example:

α = 0.05

Entire 5% rejection region is in **one side** of the distribution.

---

### Two-Tailed Test

The significance level is **split between both tails**.

Example:

α = 0.05

Left tail → 0.025  
Right tail → 0.025

---

# Advantages and Disadvantages

### Two-Tailed Test

Advantages:

- Detects effects in **both directions**
- More **conservative**

Disadvantages:

- Requires stronger evidence to reject H₀
- May increase risk of **Type II error**

---

### One-Tailed Test

Advantages:

- **More powerful** when direction is known
- Easier to detect an effect in that direction

Disadvantages:

- Cannot detect effects in the **opposite direction**
- Can lead to incorrect conclusions if direction is wrongly assumed

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4,4,500)
y = norm.pdf(x)

plt.figure(figsize=(10,5))

# Two-tailed rejection regions
z_two = 1.96
left = np.linspace(-4,-z_two,200)
right = np.linspace(z_two,4,200)

plt.plot(x,y,label="Normal Distribution")

plt.fill_between(left,norm.pdf(left),alpha=0.3,
                 label="Two-tailed rejection region (α/2)")
plt.fill_between(right,norm.pdf(right),alpha=0.3)

# One-tailed rejection region
z_one = 1.645
right_one = np.linspace(z_one,4,200)

plt.fill_between(right_one,norm.pdf(right_one),alpha=0.5,
                 label="One-tailed rejection region (α)")

plt.axvline(z_two,linestyle="--",label="Two-tailed critical ±1.96")
plt.axvline(-z_two,linestyle="--")
plt.axvline(z_one,linestyle=":",label="One-tailed critical 1.645")

plt.xlabel("Z value")
plt.ylabel("Probability Density")
plt.title("One-Tailed vs Two-Tailed Hypothesis Tests")
plt.legend()
plt.show()
````

# Choosing Between One-Tailed and Two-Tailed Tests

When performing hypothesis testing, the choice between a **one-tailed test** and a **two-tailed test** depends on the **research question**.

Use a **two-tailed test** when you want to detect a change in **either direction**.

Example:

Interpretation:

Characteristics:

- Rejection regions appear in **both tails**
- Significance level is divided between the two tails

Example:

- Left tail = 0.025
- Right tail = 0.025

# One-Tailed Test

Use a **one-tailed test** when the research hypothesis is **directional**.

Examples:

Right-tailed test

H₀: μ = 50  
H₁: μ > 50

> Testing whether a training program **increases productivity**.

Left-tailed test

H₀: μ = 50

> Testing whether a manufacturing process **reduces product weight**.

- Entire significance level is placed in **one tail**

Example:

If α = 0.05

Critical value:

Z = 1.645 (right tail)

---

# Decision Guideline

Use **two-tailed test** when:

- The direction of effect is **unknown**
- Any difference is important

Use **one-tailed test** when:

- There is a **strong theoretical reason**
- Only **one direction matters**

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4,4,500)
y = norm.pdf(x)

fig, ax = plt.subplots(1,2,figsize=(12,4))

# Two-tailed test
z_two = 1.96
ax[0].plot(x,y)
left = np.linspace(-4,-z_two,200)
right = np.linspace(z_two,4,200)

ax[0].fill_between(left,norm.pdf(left),alpha=0.3)
ax[0].fill_between(right,norm.pdf(right),alpha=0.3)

ax[0].axvline(-z_two,linestyle="--")
ax[0].axvline(z_two,linestyle="--")

ax[0].set_title("Two-Tailed Test (α split into two tails)")
ax[0].set_xlabel("Z value")
ax[0].set_ylabel("Probability Density")

# One-tailed test
z_one = 1.645
right_one = np.linspace(z_one,4,200)

ax[1].plot(x,y)
ax[1].fill_between(right_one,norm.pdf(right_one),alpha=0.3)

ax[1].axvline(z_one,linestyle="--")

ax[1].set_title("One-Tailed Test (α in one tail)")
ax[1].set_xlabel("Z value")

plt.show()
````

# Understanding P-value

## 1. Definition

The **p-value** is the probability of obtaining a result **at least as extreme as the observed sample result**, assuming the **null hypothesis (H₀) is true**.

In simple words:

**P-value measures how strong the evidence is against the null hypothesis.**

- Small p-value → Strong evidence against H₀
- Large p-value → Weak evidence against H₀

---

## 2. Interpretation

Typical interpretation using significance level α = 0.05

| p-value | Interpretation |
|-------|----------------|
| p < 0.01 | Very strong evidence against H₀ |
| 0.01 ≤ p < 0.05 | Moderate evidence against H₀ |
| 0.05 ≤ p < 0.10 | Weak evidence |
| p ≥ 0.10 | No evidence against H₀ |

Decision rule:

If  

p-value < α  

Reject H₀

Otherwise  

Fail to reject H₀

---

# Example

Suppose a company claims:

Average productivity = **50 units**

After training, a sample of employees shows:

Sample mean = **53**

Population standard deviation = **4**

Sample size = **30**

We test:

H₀: μ = 50  
H₁: μ > 50

The p-value tells us the probability of observing a mean **≥ 53** if the true mean is **50**.

If this probability is very small, the training program likely improved productivity.

---

# Interactive Visualization of P-value

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import ipywidgets as widgets

def plot_pvalue(z_score):

    x = np.linspace(-4,4,500)
    y = norm.pdf(x)

    plt.figure(figsize=(8,5))
    plt.plot(x,y)

    # shading p-value region
    x_fill = np.linspace(z_score,4,200)
    y_fill = norm.pdf(x_fill)

    plt.fill_between(x_fill,y_fill,color="red",alpha=0.4,label="p-value region")

    p_value = 1 - norm.cdf(z_score)

    plt.title(f"P-value Visualization (Z = {z_score:.2f})")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()

    plt.show()

    print("P-value =", round(p_value,5))


widgets.interact(plot_pvalue, z_score=widgets.FloatSlider(min=-3,max=3,step=0.1,value=1.5))
````

# Coin Toss Simulation to Understand P-value

## Intuition

Suppose we want to test whether a coin is **fair**.

A fair coin means:

H₀: p = 0.5  
(The probability of head is 0.5)

Alternative hypothesis:

H₁: p > 0.5  
(The coin may be biased toward heads)

---

## Experiment

We toss the coin **100 times** and observe:

Observed heads = **53**

Now we ask:

> If the coin were truly fair, what is the probability of getting **53 or more heads**?

This probability is called the **p-value**.

So the p-value measures:

**How likely our observed result is under the null hypothesis.**

---

## Simulation Idea

Instead of computing using formulas, we simulate the experiment many times:

1. Assume coin is fair
2. Simulate coin tosses many times
3. Count how often we get results **≥ observed heads**
4. That proportion is the **p-value**

---

## Interactive Simulation

````python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

def coin_pvalue_simulation(tosses=100, observed_heads=53, simulations=5000):

    results = []

    for _ in range(simulations):
        heads = np.random.binomial(tosses,0.5)
        results.append(heads)

    results = np.array(results)

    p_value = np.sum(results >= observed_heads) / simulations

    plt.figure(figsize=(9,5))
    plt.hist(results,bins=20,alpha=0.7)

    plt.axvline(observed_heads,color='red',linewidth=3,label="Observed Heads")

    plt.title("Coin Toss Simulation (Null Hypothesis: Fair Coin)")
    plt.xlabel("Number of Heads")
    plt.ylabel("Frequency")
    plt.legend()

    plt.show()

    print("Observed Heads:",observed_heads)
    print("Estimated P-value:",round(p_value,5))

    if p_value < 0.05:
        print("Decision: Reject Null Hypothesis (coin may be biased)")
    else:
        print("Decision: Fail to Reject Null Hypothesis")

widgets.interact(
    coin_pvalue_simulation,
    tosses=widgets.IntSlider(min=20,max=200,step=10,value=100),
    observed_heads=widgets.IntSlider(min=40,max=120,step=1,value=53),
    simulations=widgets.IntSlider(min=1000,max=20000,step=1000,value=5000)
)
````

# P-value in Context of Z-test (Example 1)

## Problem Statement

A company wants to check whether a new training program has increased productivity.

Given:

- Population mean (μ) = 50  
- Sample mean (x̄) = 53  
- Population standard deviation (σ) = 4  
- Sample size (n) = 30  
- Significance level (α) = 0.05  

---

## Step 1: Define Hypotheses

H₀: μ = 50  
H₁: μ > 50   (Right-tailed test)

---

## Step 2: Compute Z-statistic

Formula:

Z = (x̄ - μ) / (σ / √n)

---

## Step 3: Two Approaches

### Approach 1: Rejection Region

- Find critical Z value using α
- Compare Z-stat with critical value

### Approach 2: P-value

- Compute probability of observing such Z or more extreme
- Compare with α

---

## Let’s Solve Using Python

````python
import numpy as np
from scipy.stats import norm

# Given values
mu = 50
x_bar = 53
sigma = 4
n = 30
alpha = 0.05

# Step 1: Z-statistic
z = (x_bar - mu) / (sigma / np.sqrt(n))

# Step 2: Critical value (right-tailed test)
z_critical = norm.ppf(1 - alpha)

# Step 3: P-value
p_value = 1 - norm.cdf(z)

# Print results
print("Z-statistic:", round(z,4))
print("Critical Z value:", round(z_critical,4))
print("P-value:", round(p_value,6))
print("Alpha:", alpha)

print("\n--- Rejection Region Approach ---")
if z > z_critical:
    print("Z > Z_critical → Reject H₀")
else:
    print("Z ≤ Z_critical → Fail to Reject H₀")

print("\n--- P-value Approach ---")
if p_value < alpha:
    print("p-value < α → Reject H₀")
else:
    print("p-value ≥ α → Fail to Reject H₀")

print("\n--- Final Conclusion ---")
if (z > z_critical) and (p_value < alpha):
    print("Both methods agree: Training program significantly increased productivity.")
else:
    print("Methods do not agree (check calculations).")
````

# Key Insight

- Rejection Region checks **Z-stat vs threshold**
- P-value checks **probability vs α**

### Important:

Both methods ALWAYS give the same conclusion.

---

## Interpretation

Since:

- Z is very large (far in tail)
- p-value is very small (< 0.05)

We conclude:

✅ Strong evidence against H₀  
✅ Training program improved productivity  

---

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given
z_observed = 1.70
alpha = 0.05

# Critical value
z_critical = norm.ppf(1 - alpha)

# X values
x = np.linspace(-4, 4, 500)
y = norm.pdf(x)

plt.figure(figsize=(10,5))
plt.plot(x, y)

# Rejection region (alpha)
x_reject = np.linspace(z_critical, 4, 200)
y_reject = norm.pdf(x_reject)
plt.fill_between(x_reject, y_reject, alpha=0.4, label="Rejection Region (α)")

# P-value region
x_p = np.linspace(z_observed, 4, 200)
y_p = norm.pdf(x_p)
plt.fill_between(x_p, y_p, alpha=0.6, label="P-value Region")

# Lines
plt.axvline(z_observed, linestyle='--', linewidth=2, label="Observed Z = 1.70")
plt.axvline(z_critical, linestyle='--', linewidth=2, label="Critical Z = 1.645")

# Labels
plt.title("P-value vs Rejection Region")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.legend()

plt.show()

# Calculate p-value
p_value = 1 - norm.cdf(z_observed)

print("Z observed:", z_observed)
print("Critical Z:", round(z_critical,3))
print("P-value:", round(p_value,5))
print("Alpha:", alpha)

if p_value < alpha:
    print("\nConclusion: Reject Null Hypothesis")
else:
    print("\nConclusion: Fail to Reject Null Hypothesis")
````

# P-value in Context of Z-test (Example 2 – Two-Tailed Test)

## Problem Statement

A snack company claims:

Average weight of packets = **50 grams**

A consumer group tests this claim.

Sample data:

- Sample mean (x̄) = 49  
- Population standard deviation (σ) = 5  
- Sample size (n) = 40  
- Significance level (α) = 0.05  

---

## Step 1: Hypotheses

H₀: μ = 50  
H₁: μ ≠ 50   (Two-tailed test)

---

## Step 2: Z-statistic

Formula:

Z = (x̄ - μ) / (σ / √n)

---

## Step 3: Approaches

### Rejection Region (Two-tailed)

- α = 0.05 → each tail = 0.025  
- Critical Z = ±1.96  

Reject H₀ if:

Z < -1.96  OR  Z > 1.96  

---

### P-value Approach

P-value = 2 × P(Z ≥ |observed Z|)

---

## Let’s Solve Using Python

````python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Given values
mu = 50
x_bar = 49
sigma = 5
n = 40
alpha = 0.05

# Z-statistic
z = (x_bar - mu) / (sigma / np.sqrt(n))

# Critical values (two-tailed)
z_critical = norm.ppf(1 - alpha/2)

# P-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z-statistic:", round(z,4))
print("Critical Z values: ±", round(z_critical,4))
print("P-value:", round(p_value,6))
print("Alpha:", alpha)

print("\n--- Rejection Region Approach ---")
if abs(z) > z_critical:
    print("Reject H₀")
else:
    print("Fail to Reject H₀")

print("\n--- P-value Approach ---")
if p_value < alpha:
    print("Reject H₀")
else:
    print("Fail to Reject H₀")

print("\n--- Final Conclusion ---")
print("Both methods give the same decision.")
````

# Visualization: Two-Tailed Test

````python
# Visualization
x = np.linspace(-4,4,500)
y = norm.pdf(x)

plt.figure(figsize=(10,5))
plt.plot(x,y)

# Rejection regions
x_left = np.linspace(-4,-z_critical,200)
x_right = np.linspace(z_critical,4,200)

plt.fill_between(x_left, norm.pdf(x_left), alpha=0.3, label="Rejection Region")
plt.fill_between(x_right, norm.pdf(x_right), alpha=0.3)

# P-value regions
x_p_left = np.linspace(-4,z,200)
x_p_right = np.linspace(abs(z),4,200)

plt.fill_between(x_p_left, norm.pdf(x_p_left), alpha=0.5, label="P-value Region")
plt.fill_between(x_p_right, norm.pdf(x_p_right), alpha=0.5)

# Lines
plt.axvline(z, linestyle='--', label="Observed Z")
plt.axvline(-z_critical, linestyle='--', label="Critical Z")
plt.axvline(z_critical, linestyle='--')

plt.title("Two-Tailed Test: P-value vs Rejection Region")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.legend()

plt.show()
````

# Key Insights

- We check **both sides of distribution**
- Because deviation can be **above or below**

---

## 2. Interpretation

- If Z is far from 0 → strong evidence  
- If Z is close to 0 → weak evidence  

---

## 3. P-value Meaning

P-value represents:

Probability of observing results **as extreme in both directions**

---

## Final Conclusion

If:

p-value < α  

→ Reject H₀  

Otherwise:

→ Fail to reject H₀  

---


- One-tailed → single side  
- Two-tailed → both sides

# T-Tests in Hypothesis Testing

## What is a T-test?

A **t-test** is a statistical test used to:

- Compare the mean of a sample with a population mean, OR  
- Compare the means of two samples  

It is based on the **t-distribution**.

---

## When do we use a T-test?

We use a t-test when:

- Population standard deviation (σ) is **unknown**
- Sample size is **small**

---

## Types of T-tests

### 1. One-Sample T-test

Used to compare:

Sample mean vs Known population mean

#### Hypotheses:

H₀: μ = μ₀  
H₁: μ ≠ μ₀ (or >, < depending on test)

---

### 2. Independent Two-Sample T-test

Used to compare:

Mean of **two independent groups**

Examples:

- Marks of two different classes  
- Performance of two different algorithms  

#### Hypotheses:

H₀: μ₁ = μ₂  
H₁: μ₁ ≠ μ₂  

---

### 3. Paired T-test (Dependent T-test)

Used when samples are **related or paired**

Examples:

- Before vs After study  
- Same students' marks before and after training  

#### Hypotheses:

H₀: Mean difference = 0  
H₁: Mean difference ≠ 0  

---

## Key Idea

| Test Type | Use Case |
|----------|---------|
| One-sample | Compare sample with population |
| Independent | Compare two unrelated groups |
| Paired | Compare related measurements |

---

## Important Note

T-test uses **t-distribution instead of normal distribution** because:

- σ is unknown  
- Extra uncertainty is handled using degrees of freedom  

---

## Intuition

- Z-test → when σ is known  
- T-test → when σ is unknown  

👉 In real-world scenarios, **t-test is used more frequently**


# One-Sample T-test (Complete Example)

## Problem Statement

A manufacturer claims:

Average weight of chocolate bars = **50 grams**

We want to test this claim.

Sample data:

- Sample size (n) = 25  
- Sample mean (x̄) = 49.7 grams  
- Sample standard deviation (s) = 1.2 grams  
- Significance level (α) = 0.05  

---

## Step 1: Hypotheses

H₀: μ = 50  
H₁: μ ≠ 50   (Two-tailed test)

---

## Step 2: Why T-test?

- Population standard deviation is **unknown**
- Sample size is **small (n = 25)**

👉 Therefore, we use **t-test**

---

## Step 3: Test Statistic

Formula:

t = (x̄ - μ) / (s / √n)

---

## Step 4: Degrees of Freedom

df = n - 1 = 24

---

## Step 5: Rejection Region

For α = 0.05 (two-tailed):

t_critical ≈ ±2.064

Reject H₀ if:

t < -2.064  OR  t > 2.064

---

## Step 6: P-value Approach

P-value = 2 × P(T ≥ |t|)

---

## Solve Using Python

````python
import numpy as np
from scipy.stats import t

# Given data
mu = 50
x_bar = 49.7
s = 1.2
n = 25
alpha = 0.05

# t-statistic
t_stat = (x_bar - mu) / (s / np.sqrt(n))

# degrees of freedom
df = n - 1

# critical value
t_critical = t.ppf(1 - alpha/2, df)

# p-value
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

print("t-statistic:", round(t_stat,4))
print("Degrees of freedom:", df)
print("Critical t:", round(t_critical,4))
print("P-value:", round(p_value,6))

print("\n--- Rejection Region Approach ---")
if abs(t_stat) > t_critical:
    print("Reject H₀")
else:
    print("Fail to Reject H₀")

print("\n--- P-value Approach ---")
if p_value < alpha:
    print("Reject H₀")
else:
    print("Fail to Reject H₀")

print("\n--- Final Conclusion ---")
print("Both methods give the same result.")
````

# Visualization (t-distribution)

````python
import matplotlib.pyplot as plt
from scipy.stats import t

x = np.linspace(-4,4,500)
y = t.pdf(x, df)

plt.figure(figsize=(10,5))
plt.plot(x,y,label="t-distribution")

# rejection regions
x_left = np.linspace(-4,-t_critical,200)
x_right = np.linspace(t_critical,4,200)

plt.fill_between(x_left, t.pdf(x_left, df), alpha=0.3, label="Rejection Region")
plt.fill_between(x_right, t.pdf(x_right, df), alpha=0.3)

# p-value regions
x_p_left = np.linspace(-4,t_stat,200)
x_p_right = np.linspace(abs(t_stat),4,200)

plt.fill_between(x_p_left, t.pdf(x_p_left, df), alpha=0.5, label="P-value Region")
plt.fill_between(x_p_right, t.pdf(x_p_right, df), alpha=0.5)

# lines
plt.axvline(t_stat, linestyle='--', label="Observed t")
plt.axvline(-t_critical, linestyle='--', label="Critical t")
plt.axvline(t_critical, linestyle='--')

plt.title("One-Sample t-test Visualization")
plt.xlabel("t value")
plt.ylabel("Density")
plt.legend()

plt.show()
````

# Interpretation

- If p-value < 0.05 → Reject H₀  
- Otherwise → Fail to reject H₀  

---

## Key Insight

- t-test accounts for **extra uncertainty** due to unknown σ  
- Shape of t-distribution depends on **degrees of freedom**  

---

## Final Conclusion

Based on the result:

👉 There is / is not sufficient evidence to reject the manufacturer’s claim.

(Students should interpret based on computed values)

---

## Teaching Tip

- Z-test → σ known  
- T-test → σ unknown

# Independent Two-Sample T-test

## What is it?

An **independent two-sample t-test** is used to compare the means of **two independent groups**.

👉 Goal: Check whether the difference between two sample means is statistically significant.

---

## Example (From Notes)

A website owner wants to test:

Is there a difference in average time spent between:

- Desktop users  
- Mobile users  

---

## Given Data

Desktop users:
- n₁ = 30  
- mean₁ = 18.5 minutes  
- std₁ = 3.5  

Mobile users:
- n₂ = 30  
- mean₂ = 14.3 minutes  
- std₂ = 2.7  

Significance level:

α = 0.05  

---

## Step 1: Hypotheses

H₀: μ₁ = μ₂  
H₁: μ₁ ≠ μ₂  (Two-tailed test)

---

## Step 2: Assumptions

1. Independence of observations  
2. Normality (approximate)  
3. Equal variances (or use Welch’s test)  
4. Random sampling  

---

## Step 3: Test Statistic

For equal variance case:

t = (x̄₁ - x̄₂) / √[ s_p² (1/n₁ + 1/n₂) ]

Where:

s_p² = pooled variance

---

## Step 4: Solve Using Python

````python
import numpy as np
from scipy.stats import ttest_ind

# Data (from notes)
desktop = [12, 15, 18, 16, 20, 17, 14, 22, 19, 21, 23, 18, 25, 17, 16, 24, 20, 19, 22, 18, 15, 14, 23, 16, 12, 21, 19, 17, 20, 14]

mobile = [10, 12, 14, 13, 16, 15, 11, 17, 14, 16, 18, 14, 20, 15, 14, 19, 16, 15, 17, 14, 12, 11, 18, 15, 10, 16, 15, 13, 16, 11]

# t-test (independent samples)
t_stat, p_value = ttest_ind(desktop, mobile)

print("t-statistic:", round(t_stat,4))
print("p-value:", round(p_value,6))
````

# Step 5: Decision

````python
alpha = 0.05

if p_value < alpha:
    print("Reject H₀ → Significant difference between groups")
else:
    print("Fail to Reject H₀ → No significant difference")
````

# Paired Two-Sample T-test (Dependent T-test)

## What is a Paired T-test?

A **paired t-test** is used when:

👉 The two samples are **related or dependent**

Examples:

- Before vs After measurements  
- Same students' marks before and after training  
- Weight before and after diet program  

---

## Key Idea

Instead of comparing two separate groups:

👉 We compare **differences within the same group**

---

## Example (From Notes)

A fitness center wants to test:

👉 Does an 8-week program reduce weight?

Data:

Before:
[80, 92, 75, 68, 85, 78, 73, 90, 70, 88, 76, 84, 82, 77, 91]

After:
[78, 93, 81, 67, 88, 76, 74, 91, 69, 88, 77, 81, 80, 79, 88]

Significance level:

α = 0.05  

---

## Step 1: Hypotheses

H₀: Mean difference = 0  
H₁: Mean difference ≠ 0  

---

## Step 2: Convert to Differences

Instead of analyzing two lists:

👉 Compute:

Difference = Before - After

````python
before = [80, 92, 75, 68, 85, 78, 73, 90, 70, 88, 76, 84, 82, 77, 91]
after  = [78, 93, 81, 67, 88, 76, 74, 91, 69, 88, 77, 81, 80, 79, 88]

import numpy as np

diff = np.array(before) - np.array(after)

print("Differences:", diff)
print("Mean difference:", round(np.mean(diff),2))
````

````python
from scipy.stats import ttest_rel

t_stat, p_value = ttest_rel(before, after)

print("t-statistic:", round(t_stat,4))
print("p-value:", round(p_value,6))
````
