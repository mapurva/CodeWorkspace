# Intuition Behind Hypothesis Testing

Before learning formulas, let's understand **why hypothesis testing is needed**.

---

## Test Case 1: Is the Professor Late Today?

Imagine this situation.

Your class normally starts at **10:00 AM**.

But students have observed something over the semester:
- The professor usually arrives around **10:00 AM**
- Sometimes **1–2 minutes late**
- Rarely **more than 5 minutes late**

Today the professor arrives at **10:07 AM**.

Now students start discussing:

- Student A: "Maybe this is normal variation."
- Student B: "No, this is unusual. Something must have happened."

So the real question becomes:

> Is **10:07 AM** still part of normal variation, or is it **significantly different** from usual behavior?

---

## What Hypothesis Testing Does

Hypothesis testing gives a **systematic method** to answer this question.

We start with an assumption:

**Null Hypothesis (H₀):**
> Nothing unusual happened. The professor is arriving at the usual time.

Then we ask:

> If the professor normally arrives around 10:00 AM,  
> how likely is it to observe **10:07 AM**?

If the probability is **very small**, we say:

➡ Something unusual happened  
➡ Reject the null hypothesis

---

### Key Intuition

Hypothesis testing answers a simple question:

> **Is the observed result just random variation, or is it evidence of something new?**

# What is Hypothesis Testing?

Hypothesis testing is a **statistical method used to make decisions using data**.

In many real-world situations, we observe some data and want to answer a question:

> Is this result due to **random variation**, or does it indicate **something significant**?

Hypothesis testing provides a **structured framework** to answer this question.

---

# Basic Idea

The process begins with an assumption about the population.

This assumption is called the **Null Hypothesis**.

### Null Hypothesis (H₀)

The null hypothesis assumes that:

> **Nothing unusual is happening.**

It represents the **status quo** or the **default belief**.

Example:
- A training program does not improve productivity
- A new teaching method does not improve scores
- The average packet weight is exactly 50g

---

### Alternative Hypothesis (H₁ or Hₐ)

The alternative hypothesis represents the **claim we want to test**.

It suggests that:

> **Something has changed or there is a real effect.**

Examples:
- The training program increases productivity
- The new teaching method improves scores
- The packet weight is different from 50g

---

# What Hypothesis Testing Actually Does

Hypothesis testing evaluates the following question:

> If the null hypothesis is true, how likely is the observed data?

Two possibilities arise:

### Case 1: Data is Likely Under H₀
If the observed data is **reasonably likely**, we say:

➡ **Fail to reject the null hypothesis**

This means:
- There is **not enough evidence** to support the alternative hypothesis.

---

### Case 2: Data is Very Unlikely Under H₀
If the observed data is **very unlikely**, we say:

➡ **Reject the null hypothesis**

This means:
- There is **strong evidence** supporting the alternative hypothesis.

---

# Simple Analogy (Courtroom Example)

Hypothesis testing is similar to a **court trial**.

- **Null Hypothesis (H₀):** The defendant is innocent.
- **Alternative Hypothesis (H₁):** The defendant is guilty.

We assume innocence unless there is **strong evidence**.

If evidence is strong enough:

➡ We **reject innocence** (reject H₀)

If evidence is not strong enough:

➡ We **fail to reject innocence**

Hypothesis testing helps us answer an important question:

> **Is the observed result just random variation, or is it statistically significant evidence of change?**

# Important Points About Hypothesis Testing

Before performing any hypothesis test, it is important to understand some key principles.

---

## 1. Null Hypothesis Represents the Status Quo

The **Null Hypothesis (H₀)** usually represents the assumption that:

> Nothing new is happening.

It assumes that there is **no significant effect or relationship** between the variables being studied.

Examples:
- A new teaching method does **not** improve scores.
- A training program does **not** increase productivity.
- The average packet weight is **50 grams**.

---

## 2. Alternative Hypothesis Represents the Research Claim

The **Alternative Hypothesis (H₁ or Hₐ)** contradicts the null hypothesis.

It represents the claim that:

> There **is a significant effect or difference**.

Examples:
- The new teaching method **improves scores**.
- The training program **increases productivity**.
- The average packet weight is **different from 50 grams**.

---

## 3. We Try to Find Evidence Against the Null Hypothesis

Hypothesis testing works by using **sample data** to evaluate whether the null hypothesis is reasonable.

The goal is to determine:

> Is the observed result likely to occur if the null hypothesis were true?

If the observed data is **very unlikely under H₀**, we reject the null hypothesis.

---

## 4. Two Possible Decisions

After performing a hypothesis test, there are only two possible outcomes:

1️⃣ **Reject the Null Hypothesis (H₀)**  
This means the sample provides **sufficient evidence** supporting the alternative hypothesis.

2️⃣ **Fail to Reject the Null Hypothesis (H₀)**  
This means the sample does **not provide enough evidence** to support the alternative hypothesis.

---

## 5. Failing to Reject H₀ Does NOT Mean H₀ is True

A very important point:

> **Failing to reject the null hypothesis does not mean the null hypothesis is true.**

It only means that **the available evidence is insufficient to reject it**.

This happens because conclusions are based on **sample data**, which always contains some uncertainty.

---

## 6. Hypothesis Testing is Evidence-Based Decision Making

Hypothesis testing helps answer the question:

> Is the observed result just **random variation**, or is it **statistically significant evidence** of a real effect?

By using statistical tests, we make **data-driven decisions instead of relying on intuition alone**.

# Steps Involved in Hypothesis Testing (Rejection Region Approach)

Hypothesis testing follows a systematic procedure to decide whether the **sample data provides enough evidence to reject the null hypothesis**.

Below are the steps involved in the **Rejection Region Approach**.

---

## Step 1: Formulate the Hypotheses

Define two hypotheses:

**Null Hypothesis (H₀):**  
Represents the default assumption or status quo.  
It usually states that **there is no effect or no difference**.

**Alternative Hypothesis (H₁ or Hₐ):**  
Represents the research claim.  
It states that **there is an effect or a difference**.

Example:  
H₀: μ = 50  
H₁: μ ≠ 50

---

## Step 2: Select the Significance Level (α)

Choose a significance level **α (alpha)**, which represents the probability of committing a **Type I error**.

A **Type I error** occurs when we **reject the null hypothesis even though it is actually true**.

Common values for α:
- 0.05 (5%)
- 0.01 (1%)

---

## Step 3: Check Assumptions

Before performing the test, verify the assumptions required for the chosen statistical test.

Examples:
- Data distribution (often normal distribution)
- Independence of observations
- Sample size requirements

---

## Step 4: Choose the Appropriate Statistical Test

Depending on the problem and data characteristics, choose the appropriate test.

Examples:
- **Z-test** → when population standard deviation is known
- **t-test** → when population standard deviation is unknown
- **Chi-square test** → for categorical data
- **ANOVA** → for comparing means of multiple groups

---

## Step 5: Determine the Test Statistic and Critical Region

Calculate the **test statistic** based on the sample data.

Example (Z-test statistic):

Z = (X̄ − μ) / (σ / √n)

Using the chosen significance level (α), determine the **critical value(s)** from statistical tables.

These critical values define the **rejection region**.

If the test statistic falls in this region, the null hypothesis will be rejected.

---

## Step 6: Conduct the Test

Compute the value of the **test statistic** using the sample data.

Then compare this value with the **critical value(s)** determined earlier.

---

## Step 7: Make the Decision

Two outcomes are possible:

If the test statistic lies in the **rejection region**

➡ Reject the null hypothesis (H₀)

If the test statistic lies **outside the rejection region**

➡ Fail to reject the null hypothesis (H₀)

---

## Step 8: Interpret the Result

Finally, interpret the statistical conclusion in the context of the problem.

Example conclusion:

"The sample provides sufficient evidence to conclude that the training program significantly increases productivity."

or

"There is insufficient evidence to conclude that the average weight differs from 50 grams."

# Performing a Z-Test: Example 1

A company wants to evaluate whether a **new training program improves employee productivity**.

---

## Step 1: Problem Description

Before the training program:

- Average productivity = **50 units per day**

The population standard deviation is known:

- **σ = 5 units**

After implementing the training program, the company measures productivity for a **random sample of 30 employees**.

The sample statistics are:

- Sample size (n) = **30**
- Sample mean (x̄) = **53 units**

The question is:

> Has the training program significantly increased productivity?

---

## Step 2: State the Hypotheses

Since the company wants to check whether productivity **increased**, we use a **one-tailed test**.

**Null Hypothesis (H₀):**  
μ = 50  
(The training program does not increase productivity.)

**Alternative Hypothesis (H₁):**  
μ > 50  
(The training program increases productivity.)

---

## Step 3: Select the Significance Level

Assume the significance level is:

α = **0.05**

This means we allow a **5% chance of rejecting the null hypothesis when it is actually true**.

---

## Step 4: Choose the Test Statistic

Since the **population standard deviation (σ) is known**, we use a **Z-test**.

The Z-test statistic is: Z = (x̄ − μ) / (σ / √n)


where:

- x̄ = sample mean  
- μ = population mean  
- σ = population standard deviation  
- n = sample size  

---

## Step 5: Compute the Test Statistic

Substitute the values: Z = (53 − 50) / (5 / √30)

First calculate the denominator: 5 / √30 ≈ 5 / 5.477 ≈ 0.912

Now compute Z: Z = 3 / 0.912 ≈ 3.29


So the **test statistic is**

Z ≈ **3.29**

---

## Step 6: Determine the Critical Value

For a **right-tailed test** with α = 0.05,
https://www.ztable.net/
Critical Z value: Z₍critical₎ = 1.645


Decision rule:

- If **Z > 1.645 → Reject H₀**
- Otherwise → Fail to reject H₀

---

## Step 7: Make the Decision

Calculated Z value: Z = 3.29


Since: 3.29 > 1.645


The test statistic falls in the **rejection region**.

Therefore:

➡ **Reject the null hypothesis (H₀)**

---

## Step 8: Interpretation

There is **sufficient statistical evidence** at the 5% significance level to conclude that:

> The new training program significantly **increases employee productivity**.

# Visualizing the Rejection Region for the Z-Test

From the previous example we computed the test statistic:

Z = **3.29**

We also determined the **critical value** for a right-tailed test with α = 0.05:

Z₍critical₎ = **1.645**

Decision rule:

- If **Z > 1.645 → Reject H₀**
- If **Z ≤ 1.645 → Fail to reject H₀**

---

## Interpretation

The rejection region is located in the **right tail of the normal distribution**.

This region represents values that are **very unlikely if the null hypothesis is true**.

Since our calculated value:

Z = **3.29**

lies **far into the rejection region**, we reject the null hypothesis.

---

## What the Plot Will Show

In the visualization:

- The **blue curve** represents the normal distribution under H₀
- The **red shaded region** shows the rejection region
- The **vertical line** represents the calculated Z statistic (3.29)

This helps visually understand why we reject the null hypothesis.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Z values
z = np.linspace(-4, 4, 500)

# normal distribution
y = norm.pdf(z)

# critical value and test statistic
z_critical = 1.645
z_stat = 3.29

plt.figure(figsize=(8,5))
plt.plot(z, y, label="Normal Distribution")

# rejection region
z_reject = np.linspace(z_critical, 4, 200)
plt.fill_between(z_reject, norm.pdf(z_reject), alpha=0.3, label="Rejection Region (α = 0.05)")

# vertical lines
plt.axvline(z_critical, linestyle="--", label="Critical Value (1.645)")
plt.axvline(z_stat, linestyle=":", label="Test Statistic (3.29)")

plt.xlabel("Z value")
plt.ylabel("Probability Density")
plt.title("Rejection Region for Z-Test")
plt.legend()
plt.show()
```

# Performing a Z-Test: Example 2

A snack food company claims that their **Lays wafer packets contain an average weight of 50 grams per packet**.

A consumer watchdog organization wants to verify this claim.

---

## Step 1: Problem Description

The organization collects a **random sample of 40 packets** and measures their weights.

The sample statistics are:

- Claimed population mean (μ) = **50 grams**
- Sample mean (x̄) = **49 grams**
- Population standard deviation (σ) = **4 grams**
- Sample size (n) = **40**

The question is:

> Does the actual average packet weight differ from the claimed 50 grams?

---

## Step 2: State the Hypotheses

Since we want to check whether the average weight is **different** from 50 grams, we use a **two-tailed test**.

**Null Hypothesis (H₀):**  
μ = 50  
(The average packet weight is 50 grams.)

**Alternative Hypothesis (H₁):**  
μ ≠ 50  
(The average packet weight is different from 50 grams.)

---

## Step 3: Select the Significance Level

Assume the significance level:

α = **0.05**

Since this is a **two-tailed test**, the significance level is divided between both tails:

α/2 = **0.025** in each tail.

---

## Step 4: Choose the Test Statistic

Since the **population standard deviation (σ) is known**, we use a **Z-test**.

The test statistic formula is:

Z = (x̄ − μ) / (σ / √n)

where:

- x̄ = sample mean  
- μ = population mean  
- σ = population standard deviation  
- n = sample size  

---

## Step 5: Compute the Test Statistic

Substitute the values:

Z = (49 − 50) / (4 / √40)

First compute the denominator:

√40 ≈ 6.324  

4 / 6.324 ≈ **0.632**

Now compute Z:

Z = −1 / 0.632 ≈ **−1.58**

So the **test statistic is**

Z ≈ **−1.58**

---

## Step 6: Determine the Critical Values

For a **two-tailed test** with α = 0.05:
https://www.ztable.net/
Critical Z values are:

Z₍critical₎ = **±1.96**

Decision rule:

- If **Z < −1.96** or **Z > 1.96 → Reject H₀**
- Otherwise → **Fail to reject H₀**

---

## Step 7: Make the Decision

Calculated Z value:

Z = **−1.58**

Since:

−1.96 < −1.58 < 1.96

The test statistic lies **within the acceptance region**.

Therefore:

➡ **Fail to reject the null hypothesis (H₀)**

---

## Step 8: Interpretation

At the **5% significance level**, there is **not enough statistical evidence** to conclude that the average packet weight differs from **50 grams**.

This means the observed difference in the sample could be due to **random variation**.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate Z values
z = np.linspace(-4, 4, 500)
y = norm.pdf(z)

# Critical values and test statistic
z_critical_left = -1.96
z_critical_right = 1.96
z_stat = -1.58

plt.figure(figsize=(8,5))

# Plot normal distribution
plt.plot(z, y, label="Normal Distribution")

# Left rejection region
z_left = np.linspace(-4, z_critical_left, 200)
plt.fill_between(z_left, norm.pdf(z_left), alpha=0.3, label="Left Rejection Region (α/2 = 0.025)")

# Right rejection region
z_right = np.linspace(z_critical_right, 4, 200)
plt.fill_between(z_right, norm.pdf(z_right), alpha=0.3, label="Right Rejection Region (α/2 = 0.025)")

# Critical value lines
plt.axvline(z_critical_left, linestyle="--", label="Critical Value -1.96")
plt.axvline(z_critical_right, linestyle="--", label="Critical Value 1.96")

# Test statistic line
plt.axvline(z_stat, linestyle=":", label="Test Statistic Z = -1.58")

plt.xlabel("Z value")
plt.ylabel("Probability Density")
plt.title("Two-Tailed Z-Test Rejection Regions (α = 0.05)")
plt.legend()
plt.show()
```

# Rejection Region Approach in Hypothesis Testing

One way to perform hypothesis testing is the **Rejection Region Approach** (also called the **Critical Value Approach**).

This method decides whether to reject the null hypothesis by comparing the **test statistic** with a **critical value** determined from the probability distribution.

---

# Significance Level (α)

The **significance level**, denoted by **α (alpha)**, is a threshold used to decide whether the null hypothesis should be rejected.

It represents the probability of making a **Type I error**.

Type I Error:
> Rejecting the null hypothesis when it is actually true.

Typical values of α used in practice:

- α = 0.05 (5%)
- α = 0.01 (1%)

For example:

If α = 0.05, we accept a **5% chance of incorrectly rejecting the null hypothesis**.

---

# What is the Rejection Region?

The **Rejection Region (Critical Region)** is the set of values of the test statistic that lead to the rejection of the null hypothesis.

These values correspond to outcomes that are **very unlikely if the null hypothesis is true**.

Decision rule:

- If the test statistic falls in the **rejection region → Reject H₀**
- Otherwise → **Fail to reject H₀**

---

# Visual Intuition

In hypothesis testing, the rejection region lies in the **extreme tails of the probability distribution**.

For example, in a **two-tailed test with α = 0.05**:

- 2.5% probability lies in the **left tail**
- 2.5% probability lies in the **right tail**

These tails represent **rare events under the null hypothesis**.

If the test statistic falls in these extreme regions, we conclude that the observed result is **too unlikely to occur if H₀ were true**.

---

# Steps in the Rejection Region Approach

1. State the null and alternative hypotheses.
2. Choose the significance level (α).
3. Select the appropriate statistical test.
4. Determine the **critical value(s)** from statistical tables.
5. Compute the **test statistic** using sample data.
6. Compare the test statistic with the **critical value(s)**.
7. Decide whether to **reject or fail to reject H₀**.

---

# Limitation of the Rejection Region Approach

Although the rejection region method is useful for understanding hypothesis testing, it has some limitations.

### 1. Only Gives a Yes/No Decision

It only tells us whether to:

- Reject H₀
- Fail to reject H₀

It does **not indicate how strong the evidence is** against the null hypothesis.

---

### 2. Requires Looking Up Critical Values

We must use **statistical tables** to find critical values, which can be inconvenient.

---

### 3. Less Informative for Interpretation

The rejection region approach does not tell us:

> *How unlikely the observed result actually is.*

---

# Why We Use the p-Value Approach

To overcome these limitations, statisticians often use the **p-value approach**.

The **p-value** is defined as:

> The probability of obtaining a result **at least as extreme as the observed result**, assuming the null hypothesis is true.

Instead of comparing with a critical value, we compare the **p-value with α**.

Decision rule:

- If **p-value ≤ α → Reject H₀**
- If **p-value > α → Fail to reject H₀**

---

# Advantage of the p-Value Approach

The p-value provides **more information** than the rejection region approach.

For example:

- p-value = 0.04 → weak evidence against H₀
- p-value = 0.001 → very strong evidence against H₀

Thus, the p-value tells us **how strong the evidence is** against the null hypothesis.

---

# Key Takeaway

The **Rejection Region Approach** helps us understand the logic of hypothesis testing, while the **p-value approach** provides a more flexible and informative way to make statistical decisions.

# Visual Insight: Rejection Region vs p-value

To better understand hypothesis testing, we can visualize the **normal distribution under the null hypothesis (H₀)**.

Key ideas:

- The **center of the distribution** represents expected outcomes if H₀ is true.
- The **extreme tails** represent rare outcomes.

---

## Significance Level (α)

The **significance level α** defines the **rejection region**.

Example:

If α = 0.05 (two-tailed test)

- 2.5% probability lies in the **left tail**
- 2.5% probability lies in the **right tail**

These tail regions are the **rejection regions**.

If the test statistic falls in these regions:

➡ Reject H₀

---

## What the Visualization Shows

In the plot:

- Blue curve → sampling distribution under H₀
- Red shaded areas → **rejection regions (α)**
- Vertical dashed lines → **critical values**
- Black vertical line → **observed test statistic**

---

## Connecting to the p-value

The **p-value** measures how extreme the observed result is.

It represents:

> The probability of observing a result **as extreme or more extreme** than the test statistic assuming H₀ is true.

Unlike the rejection region approach, the p-value tells us **how strong the evidence is**.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate values
x = np.linspace(-4, 4, 500)
y = norm.pdf(x)

# Parameters
alpha = 0.05
z_critical = 1.96   # two-tailed critical value
z_stat = 2.5        # example observed statistic

plt.figure(figsize=(8,5))

# Plot distribution
plt.plot(x, y, label="Normal Distribution (H₀ true)")

# Rejection regions
left_tail = np.linspace(-4, -z_critical, 200)
right_tail = np.linspace(z_critical, 4, 200)

plt.fill_between(left_tail, norm.pdf(left_tail), alpha=0.3,
                 label="Rejection Region (α/2)")
plt.fill_between(right_tail, norm.pdf(right_tail), alpha=0.3)

# Critical values
plt.axvline(-z_critical, linestyle="--", label="Critical Values ±1.96")
plt.axvline(z_critical, linestyle="--")

# Test statistic
plt.axvline(z_stat, linestyle=":", label=f"Test Statistic Z = {z_stat}")

plt.xlabel("Z value")
plt.ylabel("Probability Density")
plt.title("Rejection Region in Hypothesis Testing")
plt.legend()
plt.show()

# Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print("p-value:", p_value)
```
