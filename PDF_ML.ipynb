How to use PDF in ML? Why to use PDF in ML?

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
print('DataFrame Info:')
df.info()
print('\nDataFrame Description:')
df.describe()
print('\nDataFrame Shape (rows, columns):')
print(df.shape)

print('\nRange of Input Features:')
for col in df.select_dtypes(include=['float64']).columns:
    print(f'{col}: Min = {df[col].min()}, Max = {df[col].max()}')

print('\nUnique Class Labels:')
print(df['species'].unique())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for feature in features:
    g = sns.FacetGrid(df, hue='species', height=4, aspect=1.5)
    g.map(sns.kdeplot, feature, fill=True)
    g.add_legend()
    g.set_axis_labels(feature, 'Density')
    g.set(title=f'PDF of {feature} by Species')
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for feature in features:
    g = sns.FacetGrid(df, hue='species', height=4, aspect=1.5)
    # Plot PDF as filled background
    g.map(sns.kdeplot, feature, fill=True, alpha=0.3, linewidth=0)
    # Plot CDF as a line
    g.map(sns.kdeplot, feature, cumulative=True, linestyle='--')
    g.add_legend()
    g.set_axis_labels(feature, 'Density / Cumulative Probability')
    g.set(title=f'PDF and CDF of {feature} by Species')
    plt.show()
```
