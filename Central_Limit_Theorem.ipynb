**Bernoulli Distribution**

Bernoulli distribution is a probability distribution that models a binary outcome, where the outcome can be either success (represented by the value 1) or failure (represented by the
value 0).

The Bernoulli distribution is named after the Swiss mathematician Jacob Bernoulli, who first introduced it in the late 1600s.

The Bernoulli distribution is characterized by a single parameter, which is the probability of success, denoted by p. The probability mass function (PMF) of the Bernoulli distribution is:


$$
P(X = x) = p^x (1 - p)^{1-x}, \quad x \in \{0,1\}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def bernoulli_pmf(p, x):
    """
    Calculates the Probability Mass Function (PMF) for a Bernoulli distribution.
    p: probability of success
    x: outcome (0 for failure, 1 for success)
    """
    if x == 0:
        return 1 - p
    elif x == 1:
        return p
    else:
        return 0

# Set the probability of success
p = 0.3 # You can change this value

# Possible outcomes for a Bernoulli distribution
x_values = [0, 1]

# Calculate the PMF for each outcome
pmf_values = [bernoulli_pmf(p, x) for x in x_values]

# Plot the PMF
fig, ax = plt.subplots()
ax.bar(x_values, pmf_values, width=0.4, color=['skyblue', 'lightcoral'])

ax.set_xticks(x_values)
ax.set_xlabel('Outcome (x)')
ax.set_ylabel('Probability P(X=x)')
ax.set_title(f'Bernoulli Distribution PMF (p={p})')
ax.set_ylim(0, 1)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add text labels for probabilities
for i, v in enumerate(pmf_values):
    ax.text(x_values[i], v + 0.05, f'{v:.2f}', ha='center', va='bottom')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def bernoulli_pmf(p, x):
    """
    Calculates the Probability Mass Function (PMF) for a Bernoulli distribution.
    p: probability of success
    x: outcome (0 for failure, 1 for success)
    """
    if x == 0:
        return 1 - p
    elif x == 1:
        return p
    else:
        return 0

# Define different probabilities of success to illustrate various cases
p_values = [0.1, 0.5, 0.9]
x_values = [0, 1]

# Create a figure and axes for the plot
fig, ax = plt.subplots(figsize=(8, 5))

bar_width = 0.2

# Plot PMF for each p value
for i, p in enumerate(p_values):
    pmf_values = [bernoulli_pmf(p, x) for x in x_values]

    # Offset x values for distinct bars for each p
    r1 = np.arange(len(x_values)) + i * bar_width

    ax.bar(r1, pmf_values, width=bar_width, label=f'p={p}', alpha=0.7)

    # Add text labels for probabilities
    for j, v in enumerate(pmf_values):
        ax.text(r1[j], v + 0.02, f'{v:.2f}', ha='center', va='bottom', fontsize=9)

ax.set_xlabel('Outcome (x)')
ax.set_ylabel('Probability P(X=x)')
ax.set_title('Bernoulli Distribution PMF for Different p values')
ax.set_xticks(np.arange(len(x_values)) + (len(p_values) - 1) * bar_width / 2)
ax.set_xticklabels(['0 (Failure)', '1 (Success)'])
ax.set_ylim(0, 1.1)
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.legend(title='Probability of Success (p)')

plt.tight_layout()
plt.show()
```

**Binomial Distribution**

Binomial distribution is a probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials with two possible outcomes (often called "success" and "failure"), where the probability of success is constant for each trial.

The binomial distribution is characterized by two
parameters: the number of trials n and the probability of success p.

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb # For binomial coefficient

def binomial_pmf(n, p, k):
    """
    Calculates the Probability Mass Function (PMF) for a Binomial distribution.
    n: number of trials
    p: probability of success on a single trial
    k: number of successes
    """
    if not (0 <= p <= 1):
        raise ValueError("Probability 'p' must be between 0 and 1.")
    if not (0 <= k <= n):
        return 0 # Probability is 0 if k is outside the valid range

    # Binomial coefficient C(n, k)
    coefficient = comb(n, k, exact=True)

    # PMF formula
    pmf = coefficient * (p**k) * ((1-p)**(n-k))
    return pmf

# Set parameters for the Binomial distribution
n = 10  # Number of trials
p = 0.5 # Probability of success

# Possible number of successes (k)
k_values = np.arange(0, n + 1) # k can range from 0 to n

# Calculate the PMF for each k
pmf_values = [binomial_pmf(n, p, k) for k in k_values]

# Plot the PMF
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(k_values, pmf_values, color='teal', alpha=0.7)

ax.set_xlabel('Number of Successes (k)')
ax.set_ylabel('Probability P(X=k)')
ax.set_title(f'Binomial Distribution PMF (n={n}, p={p})')
ax.set_xticks(k_values)
ax.set_ylim(0, max(pmf_values) * 1.1) # Add some padding to the y-axis limit
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add text labels for probabilities
for i, v in enumerate(pmf_values):
    if v > 0.005: # Only label probabilities that are visibly large enough
        ax.text(k_values[i], v + 0.01, f'{v:.3f}', ha='center', va='bottom', fontsize=9)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb # For binomial coefficient

def binomial_pmf(n, p, k):
    """
    Calculates the Probability Mass Function (PMF) for a Binomial distribution.
    n: number of trials
    p: probability of success on a single trial
    k: number of successes
    """
    if not (0 <= p <= 1):
        raise ValueError("Probability 'p' must be between 0 and 1.")
    if not (0 <= k <= n):
        return 0 # Probability is 0 if k is outside the valid range

    # Binomial coefficient C(n, k)
    coefficient = comb(n, k, exact=True)

    # PMF formula
    pmf = coefficient * (p**k) * ((1-p)**(n-k))
    return pmf

# Set a fixed number of trials
n = 20  # Number of trials

# Define different probabilities of success to illustrate various cases
p_values = [0.2, 0.5, 0.8]

# Possible number of successes (k)
k_values = np.arange(0, n + 1) # k can range from 0 to n

# Create a figure and axes for the plot
fig, ax = plt.subplots(figsize=(12, 7))

# Plot PMF for each p value
for p in p_values:
    pmf_values = [binomial_pmf(n, p, k) for k in k_values]
    ax.bar(k_values, pmf_values, alpha=0.6, label=f'p={p}')

ax.set_xlabel('Number of Successes (k)')
ax.set_ylabel('Probability P(X=k)')
ax.set_title(f'Binomial Distribution PMF for Different p values (n={n})')
ax.set_xticks(k_values[::2]) # Show fewer ticks for readability
ax.set_ylim(0, max(pmf_values) * 1.2) # Add some padding to the y-axis limit
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.legend(title='Probability of Success (p)')

plt.tight_layout()
plt.show()
```

Criteria for Binomail Distribution:
1. The process consists of n trials
2. Only 2 exclusive outcomes are possible, a success and a failure.
3. P(success) = p and P(failure) = 1-p and it is fixed from trial to trial
4. The trials are independent.

**Applications of Binomial Distribution:**

1. Binary classification problems: In binary classification problems, we often model the probability of an event happening as a binomial distribution.
For example, in a spam detection system, we may model the probability of an email being spam or not spam using a binomial distribution.


2. Hypothesis testing: In statistical hypothesis testing, we use the binomial distribution to calculate the probability of observing a certain number of successes in a given number of trials, assuming a null hypothesis is true. This can be used to make decisions about whether a certain hypothesis is supported by the data or not.


3. Logistic regression: Logistic regression is a popular machine learning algorithm used for classification problems. It models the probability of an event happening as a logistic function of the input variables. Since the logistic function can be viewed as a transformation of a linear combination of inputs, the output of
logistic regression can be thought of as a binomial distribution.

4. A/B testing: A/B testing is a common technique used to compare two different versions of a product, web page, or marketing campaign. In A/B testing, we randomly assign individuals to one of two groups and compare the outcomes of interest between the groups. Since the outcomes are often binary (e.g., click-through rate or conversion rate), the binomial distribution can be used to model the distribution of outcomes and test for differences between the groups.

**Sampling Distribution**

Sampling distribution is a probability distribution that describes the statistical properties of a sample statistic (such as the sample mean or sample variance) computed from multiple independent samples of the same size from a population.



**Population vs Sample**
Population means the entire group of individuals we want to study.

Example: Suppose we want to calculate the average salary of Indians.

India has about 1.4 billion people.

So the population is: All Indians and their salaries

If we could collect the salary of every single Indian, we could compute the true population mean salary.

But in reality: It is impossible, It is expensive, It takes too much time

Instead of collecting data from everyone, we collect data from a small subset of people.

Example: We randomly select: 10,000 Indians, record their salaries

This smaller group is called a sample.

From this sample we calculate the sample mean salary.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate salary data for a large population
# Using a log-normal distribution, which is typical for income data
# Parameters chosen to represent a hypothetical range (e.g., mean ~ 50,000 INR, some higher earners)
mean_log_salary = np.log(50000) # Log of the geometric mean salary
std_log_salary = 0.8 # Standard deviation of the log of salaries
num_samples = 100000 # Number of individuals in our simulated population

simulated_salaries = np.random.lognormal(mean=mean_log_salary, sigma=std_log_salary, size=num_samples)

# Plotting the salary distribution
plt.figure(figsize=(10, 6))
plt.hist(simulated_salaries, bins=100, density=True, alpha=0.7, color='skyblue', edgecolor='black')

plt.title('Simulated Salary Distribution for Indians')
plt.xlabel('Salary (INR)')
plt.ylabel('Probability Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Set x-axis limit to focus on the common salary range, cutting off extreme outliers for better visualization
plt.xlim(0, np.percentile(simulated_salaries, 99))

plt.tight_layout()
plt.show()

print(f"Simulated Mean Salary: ₹{np.mean(simulated_salaries):,.2f}")
print(f"Simulated Median Salary: ₹{np.median(simulated_salaries):,.2f}")
print(f"Simulated Standard Deviation: ₹{np.std(simulated_salaries):,.2f}")
```

**Why Sampling Distribution is important?**

Sampling distribution is important in statistics and machine learning because it allows us to estimate the variability of a sample statistic, which is useful for making inferences about the population. By analysing the properties of the sampling distribution, we can compute confidence intervals, perform hypothesis tests, and make predictions about the population based on the sample data.

Different samples give different estimates.

Example:

Sample 1 mean = ₹40,000

Sample 2 mean = ₹42,000

Sample 3 mean = ₹38,000

If we take many samples and compute their means, the distribution of those means is called the Sampling Distribution of the Mean.
