# Descriptive Statistics vs Inferential Statistics

Statistics is broadly divided into two types:

1. **Descriptive Statistics**
2. **Inferential Statistics**

---

## 1. Descriptive Statistics

Descriptive statistics are used to **summarize and describe the data that we have collected**.

They help us understand:
- What is the **average value**?
- What is the **spread of the data**?
- How are the values **distributed**?

Common descriptive measures include:
- Mean (average)
- Median
- Mode
- Range
- Variance
- Standard Deviation
- Graphs (histograms, bar charts, box plots)

### Example

Suppose a class has marks:
70, 75, 80, 85, 90


Descriptive statistics would summarize this data:

- Mean = 80  
- Highest = 90  
- Lowest = 70  
- Range = 20  

These values **describe the dataset but do not make predictions beyond it**.

---

## 2. Inferential Statistics

Inferential statistics are used to **draw conclusions about a population based on a sample**.

Instead of studying the entire population, we study a **small sample** and use statistical methods to make predictions.

Common inferential methods include:
- Hypothesis Testing
- Confidence Intervals
- Regression Analysis
- t-tests
- ANOVA

### Example

If we survey **100 voters** from a city of **1,000,000 voters**, we may conclude:

> "Approximately 60% of voters support Candidate A."

This conclusion is based on the **sample**, but it is used to infer information about the **entire population**.

---

## Simple Analogy

Think of a **basket of apples**:

- **Descriptive Statistics:** Measure the apples in your basket (average weight, size).
- **Inferential Statistics:** Use those apples to estimate the characteristics of **all apples in the orchard**.

---

## Key Differences

| Feature | Descriptive Statistics | Inferential Statistics |
|-------|------------------------|-----------------------|
| Purpose | Summarize data | Make predictions or conclusions |
| Data Used | Entire dataset | Sample data |
| Output | Mean, median, graphs | Confidence intervals, hypothesis tests |
| Scope | Only the collected data | Population beyond the sample |

---

## Summary

- **Descriptive Statistics:** Describe and summarize collected data.  
- **Inferential Statistics:** Use sample data to make conclusions about a larger population.

**Population Vs Sample**

Population: A population is the entire group or set of individuals, objects, or events that a researcher wants to study or draw conclusions about. It can be people, animals, plants, or even inanimate objects, depending on the context of the study. The population usually
represents the complete set of possible data points or observations.


Sample: A sample is a subset of the population that is selected for study. It is a smaller group that is intended to be representative of the larger population. Researchers collect data from the sample and use it to make inferences about the population as a whole. Since it is often impractical or impossible to collect data from every member of a population, samples are used as an efficient and cost-effective way to gather information.

# Parameter vs Statistic

In statistics, we often study a **large group (population)** but collect data from a **smaller group (sample)**.  
Two important terms used in this context are **Parameter** and **Statistic**.

---

## 1. Parameter

A **parameter** is a **numerical value that describes a population**.

Since it represents the entire population, it is usually **fixed but often unknown** because measuring the whole population may be difficult or expensive.

Parameters are typically represented using **Greek letters**.

Examples of parameters:
- Population mean (μ)
- Population standard deviation (σ)
- Population variance (σ²)
- Population proportion (p)

### Example

Suppose a university wants to know the **average height of all students**.

If we measure **every student in the university**, the calculated average height is a **parameter** because it describes the **entire population**.

---

## 2. Statistic

A **statistic** is a **numerical value calculated from a sample**.

Since it is based on a sample, its value **may vary depending on the sample chosen**.

Statistics are usually represented using **Latin letters**.

Examples of statistics:
- Sample mean (x̄)
- Sample standard deviation (s)
- Sample variance (s²)
- Sample proportion (p̂)

### Example

Instead of measuring all students, suppose we measure the height of **100 randomly selected students**.

The average height of these 100 students is a **statistic** because it describes the **sample**, not the entire population.

---

## Simple Analogy

Imagine a **large jar of marbles**.

- **Parameter:** Information about **all marbles in the jar**.
- **Statistic:** Information calculated from **a handful of marbles taken from the jar**.

---

## Key Differences

| Feature | Parameter | Statistic |
|-------|-----------|-----------|
| Definition | Numerical measure of a population | Numerical measure of a sample |
| Scope | Entire population | Sample from the population |
| Usually Known? | Often unknown | Calculated from data |
| Symbols | Greek letters (μ, σ, p) | Latin letters (x̄, s, p̂) |

---

## Summary

- **Parameter:** Describes the **population**.  
- **Statistic:** Describes a **sample** and is often used to estimate the parameter.

Inferential statistics relies on **probability theory** to quantify uncertainty and make reliable conclusions.

---

Population → Too large to measure completely  
Sample → Smaller subset used for analysis  

Using the **sample**, we estimate characteristics of the **population**.

# Point Estimate

A **point estimate** is a **single numerical value calculated from a sample that serves as the best estimate of an unknown population parameter**.

In statistics, it is often difficult or impossible to collect data from the entire population. Instead, we collect data from a **sample** and use it to estimate population characteristics such as the **population mean or population standard deviation**.

A point estimate gives **one best guess** for the value of a population parameter. :contentReference[oaicite:1]{index=1}

---

## Example

Suppose we want to estimate the **average age of students in a university**.

- Total students in the university (population): **10,000**
- Instead of surveying all students, we randomly select **100 students** (sample).

Assume the ages of the sampled students are used to compute the **sample mean**.

If the **sample mean age = 21 years**, then:

> **21 is the point estimate for the population mean age.**

This means we estimate that the **average age of all students in the university is approximately 21 years**.

---

## Common Point Estimates

Different statistics can act as point estimates for population parameters.

| Population Parameter | Point Estimate (from sample) |
|----------------------|-------------------------------|
| Population mean (μ) | Sample mean (x̄) |
| Population variance (σ²) | Sample variance (s²) |
| Population standard deviation (σ) | Sample standard deviation (s) |
| Population proportion (p) | Sample proportion (p̂) |

---

## Important Idea

A **point estimate gives a single value**, but it may not be perfectly accurate because it is based on a sample.

For example:

Estimated average age = **21**

But the true population mean might actually be **20.7 or 21.3**.

Because of this uncertainty, statisticians often use **confidence intervals**, which give a **range of possible values around the point estimate**.

---

## Summary

- A **point estimate** is a **single value calculated from sample data**.
- It is used as the **best guess for a population parameter**.
- Example: **Sample mean (x̄)** is the point estimate for **population mean (μ)**.
- Point estimates are often used to construct **confidence intervals**.

- Cricket scores naturally vary
- A range captures **uncertainty**
- The prediction becomes **more realistic**

Point estimates are useful, but **confidence intervals give more reliable information because they show uncertainty in the estimate.**

# Confidence Interval

A **confidence interval (CI)** is a **range of values that is likely to contain the true value of a population parameter** (such as the population mean).

When we calculate a statistic from a sample, it is only an **estimate** of the true population value.  
Because of sampling variability, the estimate may not be exactly equal to the true value.

To account for this uncertainty, we provide a **range of plausible values**, called the **confidence interval**.

A confidence interval is constructed using:

Confidence Interval = **Point Estimate ± Margin of Error**

Here:

- **Point Estimate** → value calculated from the sample (e.g., sample mean)
- **Margin of Error** → amount added and subtracted to create a range

---

## Example

Suppose we take a sample of **100 students** and compute their **average age**.

Sample mean (point estimate) = **28 years**

Instead of saying:

> The population mean age is exactly **28**

we calculate a **confidence interval**.

Example:

Confidence Interval = **28 ± 4**

So the interval becomes:

**[24 , 32]**

Interpretation:

> We are **95% confident** that the true population mean age lies between **24 and 32 years**.

---

## Confidence Level

The **confidence level** indicates how confident we are that the interval contains the true population parameter.

Common confidence levels:

- **90%**
- **95%**
- **99%**

Example interpretation:

A **95% confidence interval** means that if we repeatedly take samples and compute intervals, about **95 out of 100 intervals will contain the true population parameter**.

---

## Confidence Interval for Population Mean (σ known)

When the population standard deviation (σ) is known, the confidence interval for the population mean is calculated as:

$[
CI = \bar{x} \pm Z_{\alpha/2}\frac{\sigma}{\sqrt{n}}
$]

Where:

- $\bar{x} $ = sample mean  
- $ Z_{\alpha/2} $ = critical value from the standard normal distribution  
- $ \sigma $ = population standard deviation  
- $ n $ = sample size  



---

## Key Points

- A **point estimate** gives a single value.
- A **confidence interval** gives a **range of likely values**.
- The interval reflects the **uncertainty of sampling**.
- Larger samples generally produce **narrower (more precise) confidence intervals**.

---

## Summary

- A **confidence interval** estimates the range in which the **true population parameter lies**.
- It is constructed as:

**Point Estimate ± Margin of Error**

- The **confidence level** indicates how reliable the interval is.
- Confidence intervals are widely used in **statistics, research, data science, and machine learning**.

**Confidence Interval is created for Parameters and not statistics. Statistics help us get the confidence interval for a parameter.**

## Visualizing Confidence Intervals

Confidence intervals are often shown in **plots to represent uncertainty around estimates**.

Instead of showing only a single value (point estimate), plots can include **interval ranges** that indicate where the true population parameter is likely to lie.

Common ways to visualize confidence intervals include:

1. **Error bar plots**
2. **Mean with CI band**
3. **Regression line with confidence interval**
4. **Group comparison plots**

These visualizations help us understand:

- How reliable an estimate is
- How much variability exists in the data
- Whether differences between groups may be meaningful

```python
#Examples of CI usage
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Sample data (scores of a batter)
scores = np.array([45, 52, 38, 60, 49, 55, 41, 63, 47, 50])

n = len(scores)
mean = np.mean(scores)
std = np.std(scores, ddof=1)

# 95% confidence interval
confidence = 0.95
t_value = stats.t.ppf((1 + confidence)/2, df=n-1)
margin_error = t_value * (std / np.sqrt(n))

lower = mean - margin_error
upper = mean + margin_error

plt.figure()

plt.errorbar(1, mean, yerr=margin_error, fmt='o', capsize=5)

plt.xticks([1], ["Batter Average"])
plt.ylabel("Runs")
plt.title("Mean Score with 95% Confidence Interval")

plt.show()
```

# Confidence Interval for Population Mean (σ Known)

In some situations, we want to estimate the **population mean (μ)** using sample data.

If the **population standard deviation (σ) is known**, we can use the **Z-procedure** to construct a confidence interval.

The confidence interval is calculated using:

$[
\mu \approx \bar{x} \pm Z_{\alpha/2}\frac{\sigma}{\sqrt{n}}
$]

Where:

- $ \bar{x} $ = sample mean  
- $ Z_{\alpha/2} $ = critical value from the standard normal distribution  
- $ \sigma $ = population standard deviation  
- $ n $ = sample size  

This formula provides a **range of values that likely contains the true population mean**.

---

## Interpretation

A **95% confidence interval** means:

If we repeatedly take samples and compute the confidence interval each time,  
**about 95% of those intervals would contain the true population mean (μ)**.

The interval tells us **how uncertain our estimate is**.

## Assumptions for Using the Z-Procedure

To construct a confidence interval when σ is known, the following assumptions should hold:

### 1. Random Sampling
The data must be collected using **random sampling** so that the sample represents the population.

### 2. Known Population Standard Deviation
The population standard deviation **σ must be known**.

In practice, σ is rarely known. When it is unknown, we use the **t-distribution instead**.

### 3. Normal Distribution or Large Sample Size

The method assumes:

- The population distribution is **normal**, OR
- The sample size is **large (n ≥ 30)**

This works because of the **Central Limit Theorem**, which states that the sampling distribution of the mean approaches a normal distribution as the sample size increases.

## Example

Suppose we want to estimate the **average weight of packets produced by a factory**.

Known information:

Sample mean weight = **500 grams**  
Population standard deviation = **10 grams**  
Sample size = **100**  
Confidence level = **95%**

Step 1: Identify the Z value

For a **95% confidence level**:

$[
Z_{\alpha/2} = 1.96
$]

Step 2: Calculate the standard error

$[
SE = \frac{\sigma}{\sqrt{n}} = \frac{10}{\sqrt{100}} = 1
$]

Step 3: Compute the margin of error

$[
ME = 1.96 \times 1 = 1.96
$]

Step 4: Construct the confidence interval

$[
CI = 500 \pm 1.96
$]

So the interval becomes:

$[
(498.04 , 501.96)
$]

---

## Interpretation

We are **95% confident** that the **true population mean weight of packets lies between 498.04 g and 501.96 g**.

# Finding Z_{α/2} for a Confidence Interval

When constructing a confidence interval for a population mean (σ known), we use:

Here, **Z_{α/2}** is the critical value obtained from the **standard normal (Z) table**.

The value of **Z_{α/2} depends on the confidence level**.

To find it, we follow a systematic process using the **Z-table**.

## Steps to Find Z_{α/2} from Confidence Level

### Step 1: Identify the Confidence Level

Example:  
Confidence Level = **95%**

---

### Step 2: Convert to α (alpha)

Example:

α = 1 − 0.95 = **0.05**

---

### Step 3: Divide α into Two Tails

The normal distribution is symmetric, so the error is split equally.

This means:

- **2.5% area in the left tail**
- **2.5% area in the right tail**

---

### Step 4: Find the Cumulative Probability

Most Z-tables show **area from the left**.

So we calculate:

Example:

1 − 0.025 = **0.975**

---

### Step 5: Look Up the Z-table

Find **0.975** in the Z-table.

The corresponding Z value is:

So for a **95% confidence interval**, the critical value is **1.96**.

## Example Calculations

### 90% Confidence Interval

Step 1  
Confidence level = **0.90**

Step 2  

α = 1 − 0.90 = **0.10**

Step 3  

α/2 = **0.05**

Step 4  

1 − 0.05 = **0.95**

From the Z-table:

Z_{α/2} = **1.645**

---

### 95% Confidence Interval

α = 1 − 0.95 = **0.05**

α/2 = **0.025**

1 − 0.025 = **0.975**

Z_{α/2} = **1.96**

---

### 99% Confidence Interval

α = 1 − 0.99 = **0.01**

α/2 = **0.005**

1 − 0.005 = **0.995**

Z_{α/2} = **2.576**

## Common Z Critical Values

| Confidence Level | α | α/2 | Z_{α/2} |
|---|---|---|---|
| 90% | 0.10 | 0.05 | 1.645 |
| 95% | 0.05 | 0.025 | 1.96 |
| 99% | 0.01 | 0.005 | 2.576 |

Higher confidence levels produce **larger Z values** and therefore **wider confidence intervals**.

## Example: Interpreting a Confidence Interval

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import ipywidgets as widgets
from IPython.display import display

def ci_simulation(sample_size=10, pop_mean=50, pop_std=15, sims=100, conf=95):

    alpha = 1 - conf/100

    intervals = []
    contains_mean = []

    for _ in range(sims):
        sample = np.random.normal(pop_mean, pop_std, sample_size)

        x_bar = np.mean(sample)
        s = np.std(sample, ddof=1)

        t_val = stats.t.ppf(1 - alpha/2, df=sample_size-1)
        margin = t_val * (s/np.sqrt(sample_size))

        lower = x_bar - margin
        upper = x_bar + margin

        intervals.append((lower, upper))
        contains_mean.append(lower <= pop_mean <= upper)

    plt.figure(figsize=(10,6))

    for i,(lower,upper) in enumerate(intervals):

        if contains_mean[i]:
            color='blue'
        else:
            color='orange'

        plt.plot([i,i],[lower,upper], color=color)
        plt.scatter(i,lower,color=color,s=10)
        plt.scatter(i,upper,color=color,s=10)

    plt.axhline(pop_mean, color='red', label='Population Mean')

    plt.title("Confidence Interval Simulation")
    plt.xlabel("Simulation Number")
    plt.ylabel("Confidence Interval")

    plt.legend()
    plt.show()

widgets.interact(
    ci_simulation,
    sample_size = widgets.IntSlider(min=5,max=100,step=1,value=10,description="Sample Size"),
    pop_mean = widgets.FloatSlider(min=0,max=100,step=1,value=50,description="Population Mean"),
    pop_std = widgets.FloatSlider(min=1,max=30,step=1,value=15,description="Population Std"),
    sims = widgets.IntSlider(min=10,max=200,step=10,value=100,description="Simulations"),
    conf = widgets.IntSlider(min=80,max=99,step=1,value=95,description="Confidence %")
)
```

$$

### Meaning of Symbols

- **$\bar{x}$** : Sample mean (point estimate)  
- **$Z_{\alpha/2}$** : Critical value from the standard normal distribution  
- **$\sigma$** : Population standard deviation  
- **$n$** : Sample size  

---

The **margin of error** is the part added and subtracted from the sample mean:

$$
ME = Z_{\alpha/2}\frac{\sigma}{\sqrt{n}}
$$


From this formula, we can see that the **margin of error depends on three factors**:

1. **Confidence Level**
2. **Sample Size**
3. **Population Standard Deviation**

These factors determine how **wide or narrow** the confidence interval will be.

## 1. Confidence Level

The **confidence level** determines the value of the critical value $Z_{\alpha/2}$.

Higher confidence levels require **larger Z values**, which increases the margin of error.

Example Z values:

| Confidence Level | Z value |
|---|---|
| 90% | 1.645 |
| 95% | 1.96 |
| 99% | 2.576 |

Effect:

Higher confidence → **larger margin of error → wider interval**

Reason:

To be more confident that the interval contains the true population parameter, the interval must be wider.

## 2. Sample Size

Sample size appears in the denominator of the margin of error formula:



When the **sample size increases**, the value of $( \sqrt{n} $) increases.

This causes the **margin of error to decrease**.

Effect:

Larger sample size → **smaller margin of error → narrower interval**

This is why collecting more data improves the **precision of estimates**.

## 3. Population Standard Deviation (σ)

The population standard deviation measures **variability in the data**.

If the data values vary a lot, the estimate becomes less precise.

Effect:

Larger standard deviation → **larger margin of error → wider interval**

Smaller standard deviation → **smaller margin of error → narrower interval**

In simple terms:

More variability in the data means **greater uncertainty in the estimate**.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import ipywidgets as widgets
from IPython.display import display

def ci_calculator(confidence=95, sample_mean=50, sigma=15, n=30):

    alpha = 1 - confidence/100
    z = norm.ppf(1 - alpha/2)

    margin = z * (sigma/np.sqrt(n))

    lower = sample_mean - margin
    upper = sample_mean + margin

    print(f"Critical Value (Z-score): {round(z,3)}")
    print(f"Margin of Error: {round(margin,3)}")
    print(f"Confidence Interval: ({round(lower,3)}, {round(upper,3)})")

    x = np.linspace(sample_mean-4*sigma, sample_mean+4*sigma, 400)
    y = norm.pdf(x, sample_mean, sigma)

    plt.figure(figsize=(8,4))

    plt.plot(x,y)

    plt.axvline(lower, linestyle='--', label="Lower Bound")
    plt.axvline(upper, linestyle='--', label="Upper Bound")
    plt.axvline(sample_mean, linestyle='-', label="Sample Mean")

    plt.fill_between(x,y,where=(x>=lower)&(x<=upper),alpha=0.3)

    plt.title("Confidence Interval Visualization")
    plt.legend()

    plt.show()


widgets.interact(
    ci_calculator,
    confidence = widgets.IntSlider(min=80,max=99,step=1,value=95,description="Confidence Level"),
    sample_mean = widgets.FloatSlider(min=0,max=100,value=50,description="Sample Mean"),
    sigma = widgets.FloatSlider(min=1,max=30,value=15,description="Population Std"),
    n = widgets.IntSlider(min=5,max=200,value=30,description="Sample Size")
)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set parameters
sample_size = 30
confidence_level = 95

# Calculate the critical value for the Z-score
critical_value = stats.norm.ppf((1 + confidence_level / 100) / 2)

# Generate an array of population standard deviations
population_std_devs = np.arange(1, 101)

# Calculate the margin of error for each population standard deviation
margin_of_errors = critical_value * (population_std_devs / np.sqrt(sample_size))

# Plot the relationship
plt.plot(population_std_devs, margin_of_errors)
plt.xlabel("Population Standard Deviation")
plt.ylabel("Margin of Error")
plt.title("Relationship Between Margin of Error and Population Standard Deviation (Z Procedure)")
plt.grid(True)
plt.show()
```

# Confidence Interval using the t-Procedure

In many practical situations, the **population standard deviation (σ) is unknown**.

Instead of σ, we estimate variability using the **sample standard deviation (s)**.

Because this introduces additional uncertainty, we use the **Student’s t-distribution** instead of the standard normal distribution.

This method for constructing confidence intervals is called the **t-procedure**.

## Confidence Interval Formula (σ Unknown)

The confidence interval for the population mean is:

$$
CI = \bar{x} \pm t_{\alpha/2}\frac{s}{\sqrt{n}}
$$

Where:

- **$\bar{x}$** → sample mean  
- **$t_{\alpha/2}$** → critical value from the t-distribution  
- **$s$** → sample standard deviation  
- **$n$** → sample size  

The term

$$
t_{\alpha/2}\frac{s}{\sqrt{n}}
$$

is called the **margin of error**.

## Degrees of Freedom

The t-distribution depends on the **degrees of freedom (df)**.

$$
df = n - 1
$$

Example:

If the sample size is **20**

$$
df = 20 - 1 = 19
$$

The **t critical value** is obtained from the **t-table** using:

1. Confidence level
2. Degrees of freedom

## Assumptions for Using the t-Procedure

To apply the t-procedure, the following conditions should be satisfied:

### 1. Random Sampling

The data must be collected using **random sampling** so that the sample represents the population.

This helps reduce bias and allows conclusions about the entire population. :contentReference[oaicite:2]{index=2}

---

### 2. Population Standard Deviation Unknown

The population standard deviation **σ is unknown**, so we estimate it using the **sample standard deviation s**.

The t-distribution accounts for the additional uncertainty introduced by using **s instead of σ**. :contentReference[oaicite:3]{index=3}

---

### 3. Approximately Normal Distribution

The t-procedure assumes that the population is approximately **normally distributed**.

If the population distribution is not normal, the procedure can still be used when the **sample size is large**, due to the **Central Limit Theorem**. :contentReference[oaicite:4]{index=4}

---

### 4. Independent Observations

The observations in the sample should be **independent of each other**.

This means the value of one observation does not influence another observation.

## Normal Distribution vs Student’s t-Distribution

The **Student’s t-distribution** is used when:

- The **population standard deviation (σ) is unknown**
- We estimate variability using the **sample standard deviation (s)**
- The **sample size is small**

The t-distribution is **similar to the normal distribution**, but there is an important difference.

### Key Difference

The **t-distribution has heavier tails** than the normal distribution.

This means:

- It assigns **more probability to extreme values**
- Confidence intervals become **wider**
- This accounts for the **extra uncertainty** caused by estimating σ with s.

### Role of Degrees of Freedom

The shape of the t-distribution depends on the **degrees of freedom (df)**.

$$
df = n - 1
$$

When **df is small**:
- t-distribution has **very heavy tails**

When **df increases**:
- t-distribution becomes **closer to the normal distribution**

As the **sample size becomes large**, the **t-distribution approaches the normal distribution**.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

# x-axis values
x = np.linspace(-5, 5, 500)

# Normal distribution
normal = norm.pdf(x)

# t-distributions with different degrees of freedom
t1 = t.pdf(x, df=1)
t5 = t.pdf(x, df=5)
t10 = t.pdf(x, df=10)
t30 = t.pdf(x, df=30)

plt.figure(figsize=(8,5))

plt.plot(x, normal, label="Normal Distribution", linewidth=2)
plt.plot(x, t1, label="t (df = 1)")
plt.plot(x, t5, label="t (df = 5)")
plt.plot(x, t10, label="t (df = 10)")
plt.plot(x, t30, label="t (df = 30)")

plt.title("Normal Distribution vs Student's t-Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")

plt.legend()
plt.grid(True)

plt.show()
```

From the graph:

df = 1

very heavy tails

very wide distribution

df = 5

still wider than normal

df = 10

closer to normal

df = 30

almost identical to the normal distribution

This visually explains the key idea:

As degrees of freedom increase,
t distribution → normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t
import ipywidgets as widgets
from IPython.display import display

def compare_distributions(df=5):

    # x values
    x = np.linspace(-4,4,500)

    # distributions
    normal = norm.pdf(x)
    t_dist = t.pdf(x, df)

    plt.figure(figsize=(8,5))

    plt.plot(x, normal, label="Normal Distribution")
    plt.plot(x, t_dist, label=f"Student's t-Distribution (df={df})")

    plt.title("Normal and t-Distribution Comparison")
    plt.xlabel("Value")
    plt.ylabel("Probability Density")

    plt.legend()
    plt.grid(True)

    plt.show()


widgets.interact(
    compare_distributions,
    df=widgets.IntSlider(min=1,max=50,step=1,value=5,description="Degrees of Freedom")
)
```

## Example

Suppose a researcher collects the following sample data representing **hours spent studying per day** by 8 students:

2, 3, 4, 5, 3, 4, 6, 5

Step 1: Calculate sample mean

$$
\bar{x} = 4
$$

Step 2: Calculate sample standard deviation

$$
s \approx 1.31
$$

Step 3: Identify sample size

$$
n = 8
$$

Step 4: Calculate degrees of freedom

$$
df = n - 1 = 7
$$

Step 5: Find t critical value

For **95% confidence** and **df = 7**

$$
t_{0.025,7} \approx 2.365
$$

Step 6: Calculate margin of error

$$
ME = 2.365 \times \frac{1.31}{\sqrt{8}}
$$

$$
ME \approx 1.09
$$

Step 7: Construct confidence interval

$$
CI = 4 \pm 1.09
$$

Final interval:

$$
(2.91,\ 5.09)
$$

Interpretation:

> We are **95% confident that the true average study time lies between 2.91 and 5.09 hours**.
